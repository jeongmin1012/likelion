### 주석
- 코드는 위에서 아래 방향
- 이 과정에서 만나는 코드를 실행
- 이 때, 주석을 만나게 되면 주석은 무시
- 사람 입장에서는 코드에 대해서 기록하고 싶은 내용 (설명 등)을 작성할 때 사용
- 작성한 코드 중에서 실행하고 싶지 않은 부분을 주석 처리 하기도
- 주석 단축키 : 주석 처리하고 싶은 코드 부분을 선택한 다음 'ctrl+c/(ctrl 떼지 않고)+k'
- 주석 제거 : 주석 처리하고 싶지 않은 코드 부분을 선택한 다음 'ctrl+c/(ctrl 떼지 않고)+u'

```python
# 이 부분은 주석입니다.
# 주석은 실행 될 때 무시되는 부분입니다.

a1 = 10 + 20
print(a1)
```

## raw string
- 작성한 것 그대로 관리하는 문자열
- 문자열 안에 작은 따옴표와 큰 따옴표가 모두 있을 경우
- 여러 줄로 되어 있는 문자열을 작성하고자 할 경우
- 공백 조차도 모두 문자열에 포함되기 때문에 들여쓰기 같은 것을 할 때 주의!!!

```python
#만약 큰 따옴표를 포함하는 문자열을 사용하고 싶다면 작은 따옴표로 묶어준다.
print('이름은 "홍길동" 이고 나이는 "10살" 입니다')
#만약 작은 따옴표를 포함하는 문자열을 사용하고 싶다면 큰 따옴표로 묶어준다.
print("이름은 '홍길동'이고 나이는 '10살' 입니다")

#pyhton 문법상 오류
# print('이름은 '홍길동' 이고 나이는 '10살' 입니다')
# print("이름은 "홍길동"이고 나이는 "10살" 입니다")
```

```python
# 다음과 같이 작음 따옴표 문자열 안에 작은 따옴표를 사용하거나
# 큰 따옴표 문자열 안에 큰 따옴표를 사용하면 오류가 발생한다
print('이름은 '홍길동' 이고 나이는 '10살' 입니다')
print("이름은 "홍길동" 이고 나이는 "10살" 입니다")
#'이름은' <-여기만 주석으로 묶인 후 '이고 나이는' '입니다' 이렇게만 주석처리 됨 ~변수가 아닌 홍길동, 10살 때문에 에러
```

- 만약 작음 따옴표 문자열 안에 작은 따옴표를 사용 or  큰 따옴표 문자열 안에 큰 따옴표를 사용하고 싶다 
- ->에스케이프 문자를 사용
- \n - 밑으로 내려준다. \" - 큰 따옴표, \' - 작은 따옴표

```python
print('이름은 \'홍길동\' 이고 나이는 \'10살\'입니다')
print("이름은 \"홍길동\" 이고 나이는 \"10살\"입니다")
```

```python
# 하지만 이렇게 쓰는 것이 너무 불편하기 때문에 raw string 을 지원
print('''이름은 '홍길동' 이고 나이는 "10살"입니다''')
print("""이름은 '홍길동' 이고 나이는 "10살"입니다""")
print('''
      동해물과 백두산이
      마르고 닳도록
      하느님이 보우하사
      우리나라 만세''')
print('''동해물과 백두산이
마르고 닳도록
하느님이 보우하사
우리나라 만세''')
print('''
동해물과 백두산이
마르고 닳도록
하느님이 보우하사
우리나라 만세
      ''')
```
