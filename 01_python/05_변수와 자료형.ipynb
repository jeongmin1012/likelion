### 리터럴
- 프로그래밍 언어에서 값을 작성하는 문법

```python
print('정수 :', 100)
print('실수 :', 11.11)
print('문자열 :', '하하하')
print('참  :', True)
print('거짓 :', False)
print('지수 : ', 2e5) #10의 5제곱에 2곱한 것
print('복소수 :', 10 + 8j)
print('세자리 마다 표시 : ', 123_456_789)
print('없음 :', None)
```

```python
# type을 사용하면 값의 타입을 파악할 수 있다.
print(type(100))
print(type(11.11))
print(type('하하하'))
print(type(True))
print(type(False))
print(type(2e10))
print(type(10 + 8j))
print(type(123_456_789))
print(type(None))
```

### 변수
- 개발자가 원하는 값을 저장할 수 있는 기억공간을 의미
- 파이썬은 이름만 정해주면 변수 정의 가능

- 파이썬은 변수의 선언과 동시에 저장할 값을 지정해줘야 함
- = : 대입연산자. 오른쪽에 있는 값을 왼쪽의 변수에 저장하는 작업을 수행
- 변수에 임의의 값을 꼭 대입하여야 함. 그렇지 않을 경우 오류 발생

```python

a1 = 10
a2 = 11.11
a3 = '안녕하세요'
```

```python
# 변수의 이름만 사용하면 변수에 저장되어 있는 값을 가져와 사용할 수 있다.
print(a1)
print(a2)
print(a3)
```

- 파이썬은 변수에 다른 타입의 값을 저장하는 것이 가능
- 사실 파이썬은 개발자가 사용하는 타입을 1가지로 통일이라 볼 수 있음
- 사람이 보기에는 타입이 다른 것 같지만 실제로는 모두 '객체' 라는 이름의 타입으로 관리

```python
a4 = 100
print(a4)
print(type(a4))

a4 = 11.11
print(a4)
print(type(a4))

a4 = '문자열'
print(a4)
print(type(a4))
```

```python
# 다수의 변수에 같은 값을 모두 저장하는 방법
# 오른쪽에 있는 값을 좌측의 변수에 저장하는 작업을 하는 것 뿐
a5 = a6 = a7 = 100
#a7=10 -> a7(10)=a6 ->a6(10)=a5
#결과적으로 모두 10이 대입됨

print(a5)
print(a6)
print(a7)
```

```python
# 각 변수에 서로 다른 값을 동시에 담아주는 ㅇ방법
# 문법적으론 불가능
# 여기에는 Tuple 이용 ->추후 수업할 예정
a8, a9, a10 = 100, 200, 300

print(a8)
print(a9)
print(a10)
```

- 파이썬 3.6부터 변수에 저장할 값을 명시 가능
- 이러한 경우 변수를 정의할 때 자료형을 명시하면 값을 지정하지 않아도 됨
- 그렇지 않다면 변수를 사용하기 전에 반드시 변수에 값을 저장

```python
a11:int

a11 = 100
print(a11)

# 변수에 타입을 정했다고 하더라도 다른 타입의 값을 저장하는 것이 가능
# 이것은 파이썬에서는 모든 값이 '객체'라는 한 가지 형태로 관리되기 때문
a11 = 11.11
print(a11)
```


### 변수에 자료형을 붙혀주는 이유
1. 변수를 선언할 때 최초의 값을 지정하고 싶지 않을 경우 -> (해결방안) 그냥 변수에 0과 같은 임의의 값을 지정
2. 변수의 타입을 명시적으로 표현하고 싶을 때 -> (해결방안) 해당 변수에 다른 타입의 값을 담을 수 있기 때문에 의미 X

- 파이썬은 변수를 선언할 때 변수에 저장되는 값의 타입을 지정할 수 있지만 모든 값들이 '객체'라는 한 가지 타입으로 관리되기 때문에 변수에 자료형을 붙혀주는 것이 무의미
