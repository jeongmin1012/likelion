### 연산자
- 값 들을 가지고 계산을 하기 위한 특수 문자들
- 키보드에 있는 특수문자 대부분은 연산자에 해당한다.
- 일부 특수 기호는 의미가 없다. (식별자에 사용할 수 있다.  _, $ 등등)

```python
# 산술 연산자
v1 = 10
v2 = 20

# 더하기
a1 = v1 + v2
print('v1 + v2 =', a1)

# 빼기
a1 = v1 - v2
print('v1 - v2 =', a1)

# 곱하기
a1 = v1 * v2
print('v1 * v2 =', a1)

# 나누기 : 결과는 실수
a1 = v1 / v2
print('v1 / v2 =', a1)

# 나누기 : 결과는 정수
a1 = v1 // v2
print('v1 // v2 =', a1)

# 나머지
a1 = v1 % 3
print('v1 % 3 =', a1)

# 거듭 제곱
a1 = v1 ** 3
print('v1 ** 3 =', a1)
```

```python
# 비교 연산자
# 좌측의 값이 우측이 비해 어떠한가..
# 결과는 True나 False가 나온다.
v1 = 10
print('v1 :',v1)

# v1이 10과 같은가
print('v1 == 10 :',v1 == 10)
# v1이 20과 같은가
print('v1 == 20 :',v1 == 20)
# v1이 10과 다른가
print('v1 != 10 :',v1 != 10)
# v1이 20과 다른가
print('v1 != 20 :',v1 != 20)
# v1이 5보다 큰가
print('v1 > 5 :',v1 > 5)
# v1이 20보다 큰가
print('v1 > 20 :',v1 > 20)
# v1이 5보다 작은가
print('v1 < 5 :',v1 < 5)
# v1이 20보다 작은가
print('v1 < 20 :',v1 < 20)
# v1이 5보다 크거나 같은가
print('v1 >= 5 :',v1 >= 5)
# v1이 10보다 크거나 같은가
print('v1 >= 10 :',v1 >= 10)
# v1이 20보다 크거나 같은가
print('v1 >= 20 :',v1 >= 20)
# v1이 5보다 작거나 같은가
print('v1 <= 5 :',v1 <= 5)
# v1이 10보다 작거나 같은가
print('v1 <= 10 :',v1 <= 10)
# v1이 20보다 작거나 같은가
print('v1 <= 20 :',v1 <= 20)
```

### 대입연산자
- 우측의 값을 좌측의 변수에 담아주는 연산자.
- 연산자 우선 순위 중 가장 마지막에 수행되는 연산자.

```python
v1 = 10
v2 = 10

v1 = v1 + 3
v2 += 3

print(v1)
print(v2)

v1 = v1 - 3
v2 -= 3

print(v1)
print(v2)
```

### 논리 연산자
- 연산자 좌측과 우측에 결과가 True나 False은 연산식(비교 연산자)이 있어야 한다.
- and : 둘 중 하나라도 False면 최종 결과는 False가 된다.
- or : 둘 중 하나라도 True면 최종 결과는 True가 된다.
- not : 우측의 결과가 True면 False로 , False라면 True로 변경한다.

```python
v1 = 10
v2 = 20

print('v1 > 5 and v2 > 10 :', v1 > 5 and v2 > 10)
print('v1 > 20 and v2 > 10 :', v1 > 20 and v2 > 10)

print('v1 < 5 or v2 < 10 :',v1 < 5 or v2 < 10)
print('v1 > 5 or v2 < 10 :', v1 > 5 or v2 < 10)

print('not v1 < 5 :', not v1 < 5)
```
