### if 문
- if 문에 주어진 연산식의 결과가 True일 경우에만 관리하는 코드가 수행된다.

```python
a1 = 100
```

```python
# 현재 a1에는 100이 저장되어 있기 때문에 비교연산자를 수행한 연산의 결과는 True가 된다.
# 이에 if문이 관리하는 코드가 수행된다.
if a1 == 100 :
    print('a1은 100입니다')
```

```python
# 현재 a1에는 100이 저장되어 있기 때문에 비교연사자를 수행한 연산의 결과는 False가 된다.
# 이에 if문이 관리하는 코드가 수행되지 않는다.
if a1 != 100 :
    print('a1은 100이 아닙니다')
```

```python
# 파이썬에서는 if 문과 같이 어떠한 코드를 관리하는 요소들은 자신이 관리하는 코드의 범위를
# 들여쓰기로 구분한다.
# 파이썬 코드를 작성하실 때는 들여쓰기를 꼭~~~ 주의해주세요........(어? 소리 나옵니다.)
if a1 != 100 :
    print('if 1')
    print("이 부분이 수행될까요? 1")

if a1 != 100 :
    print('if 2')
print("이 부분이 수행될까요? 2")
```

```python
# else
# if 문의 조건식의 결과가 False인 경우 수행되는 부분
if a1 == 100 :
    print('a1은 100입니다')
else :
    print('a1은 100이 아닙니다')


if a1 != 100 :
    print('a1은 100이 아닙니다')
else :
    print('a1은 100입니다')
```

```python
# elif : 설정해야 하는 조건이 여러가지인 경우

if a1 == 50 :
    print('a1은 50입니다')
elif a1 == 70 :
    print('a1은 70입니다')
elif a1 == 100 :
    print('a1은 100입니다')
elif a1 == 150 :
    print('a1은 150입니다')
else :
    print('a1은 50, 70, 100, 150이 아닙니다')
```
