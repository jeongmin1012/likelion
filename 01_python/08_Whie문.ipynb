### While
- 주어진 조건 수식의 결과가 True인 경우 관리하는 코드를 수행하고 반복한다.

```python
a1 = 0

while a1 < 10 :
    print('while a1 :', a1)
    a1 = a1 + 1
```

- 파이썬에서 for 문은 다수의 기억장소를 관리하는 요소를 통해 반복하는 것만 지원된다.
- 뒤에서 다수의 기억장소를 관리하는 개념을 배울 것이다.

- range : 0부터 주어진 숫자 -1 까지 1씩 증가하는 값을 관리하는 요소이다.
- 지금은 그냥 반복횟수를 결정하는 용도로 사용하겠습니다.

```python
for a1 in range(10):
    print('for :',a1)
```

### break
- 반복문 내부에서 사용한다.
- 반복문 내부의 코드가 수행 중에 break를 만나면 반본문 수행이 종료된다.

### continue
- 반복문 내부에서 사용한다.
- 반복문 내부의 코드가 수행 중에 continue를 만나면 처리할 코드가 더 있다고 하더라도 다음 반복으로 넘어간다.

```python
a1 = 0

while a1 <= 10 :
    a1 = a1 + 1

    # a1이 5보다 커지면 반복을 중단한다.
    if a1 > 5 :
        break

    print(a1)
```

```python

a1 = 0

while a1 <= 10 :
    a1 =  a1 + 1

    # a1이 짝수가 아니라면 다음 반복으로 넘어간다.
    if a1 % 2 != 0 :
        continue

    print(a1)
```
