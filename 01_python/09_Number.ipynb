### Number
- 파이썬에서 사용되어지는 모든 숫자 값들은 Number 라는 형태의 요소로 관리가 된다.
- int(정수 값), floot(실수 값), bool(논리 값), abs(절댓값)

### 값의 변환

```python
# 정수 값으로 변환
a1 = int(11.11)
a2 = int(True)
a3 = int('100')
print(a1, a2, a3)
```

```python
# 실수 값으로 변환
a1 = float(11)
a2 = float(True)
a3 = float('11.11')

print(a1, a2, a3)
```

```python
# 논리 값으로 변환
a1 = bool(11)
a2 = bool(11.11)
a3 = bool('100')
a4 = bool(0)

print(a1, a2, a3, a4)
```

```python
# 절대값
# ( ) 안에 넣어준 값의 절대값을 구해준다.
a1 = abs(30)
a2 = abs(-30)
print(a1)
print(a2)
```

### 수학에 관련된 기능을 사용하고 싶을 때

```python
import math
```

- math.ceil() : 소수점 이하 올림
- math.floor() : 소수점 이하 내림
- min() : () 안 값에서 최솟값
- max() : () 안 값에서 최댓값

```python
# 소수점 이하 올림
a1 = math.ceil(100)
a2 = math.ceil(55.55)
a3 = math.ceil(-55.55)

print(a1, a2, a3)
```

```python
# 소수점 이하 내림
a1 = math.floor(100)
a2 = math.floor(55.55)
a3 = math.floor(-55.55)

print(a1, a2, a3)
```

```python
# 주어진 값에서 최소 또는 최대 값을 가져온다.
a1 = min(30, -10, -20, 40)
a2 = max(30, -10, -20, 40)

print(a1, a2)
```

- random.seed() : 매번 다른 값 나옴
- () 안에 숫자를 넣어주면 일정한 값 추출 가능
- ex) random.seed(100), random.seed(10)

```python
# 랜덤을 사용하고 싶을 때
import random

# 실행할 때 마다 같은 패턴의 랜덤값을 사용하고 싶다면..
random.seed(100)

for a1 in range(10) :
    a1 = random.random()
    print(a1)
```

----

```python
# 3 ~ 15까지의 범위에서 랜덤 값을 추출한다.

# 숫자의 개수를 구한다.
cnt = 15 - 3 + 1

# 랜덤값을 구한다 (0 ~ 0.999999999999)
r1 = random.random()
# 숫자의 개수를 곱해준다 (0 ~ 12.999999999999999)
r2 = r1 * cnt
# 최소 숫자를 더한다. (3 ~ 15.9999999999999999999)
r3 = r2 + 3
# 소수점 이하를 버린다. (3 ~ 15)
r4 = math.floor(r3)
print(r4)
```

```python
# 위에서 작성한 공식대로 랜덤값을 계산해서 전달해준다.
# random.randint(최소값, 최대값)
r1 = random.randint(3, 15)
print(r1)
```
