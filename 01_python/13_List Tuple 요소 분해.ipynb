```python
tuple1 = 10, 20, 30, 40, 50
list1 = [10, 20, 30, 40, 50]
```

```python
# 리스트나 튜블이 가지고 있는 값의 개수 만큼의 변수를 나열해 주면
# 1:1 대응하여 변수에 담아준다.
#동일한 갯수가 아니면 에러 발생

a1, a2, a3, a4, a5 = tuple1
a10, a20, a30, a40, a50 = list1

print(a1, a2, a3, a4, a5)
print(a10, a20, a30, a40, a50)
```

```python
# 100, 200, 300 : 여기서 값 3개를 가지고 있는 튜플이 만들어진다.
# 그 후 좌측에 튜플이 관리하는 값의 개수와 동일한 개수의 변수를 나열하였기 때문에
# 각각 1:1 대응하여 변수에 담게 된다.
# 마치 각 변수에 값들을 각각 집어넣는 문법이 있는 것 처럼 보이지만
# 사실 중간에 tuple을 이용하게 된다.
a8, a9, a10 = 100, 200, 300

print(a8)
print(a9)
print(a10)
```
