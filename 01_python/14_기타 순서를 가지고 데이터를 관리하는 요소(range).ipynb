```python
#0 ~ 20 -1 까지 1씩 증가하는 값을 관리하는 요소
r1 = range(20)
print(r1)
print(f'시작 값 : {r1.start}')
print(f'종료 값 : {r1.stop}')
print(f'증감 값 : {r1.step}')
```

```python
# 리스트나 튜플로 만들어서 출력해본다.
print(list(r1))
```

```python
# 3 ~ 27 - 1 까지 1씩 증가하는 range
r2 = range(3, 27)
print(list(r2))

# 3 ~ 27 - 1 까지 2씩 증가하는 range
r3 = range(3, 27, 2)
print(list(r3))
```

```python
# 30 ~ 8 + 1 까지 1씩 감소
r4 = range(30, 8, -1)
print(list(r4))
```
