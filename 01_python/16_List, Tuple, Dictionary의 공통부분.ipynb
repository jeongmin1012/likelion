```python
list1 = [10, 20, 30, 40, 50]
tuple1 = 10, 20, 30, 40, 50
dict1 = {
    'a1' : 100,
    'a2' : 200,
    'a3' : 300
}
```

```python
# in
# 튜플, 리스트 : 주어진 값이 포함되어 있는 가...
# 딕셔너리 : 주어진 이름으로 저장되어 있는 값이 있는 가..
print(10 in list1)
print(100 in list1)
print(10 in tuple1)
print(100 in tuple1)
print('a1' in dict1)
print('a100' in dict1)
```

```python
# enumerate
# 리스트와 튜플만 사용할 수 있다.
e1 = enumerate(list1)
e2 = enumerate(tuple1)

print(e1)
print(e2)

# enumerate
# 리스트와 튜플만 사용할 수 있다.
# 0부터 1씩 증가되는 순서 값과 해당 번째의 값을 묶은 튜플들이 만들어진다.
e1 = enumerate(list1)
e2 = enumerate(tuple1)

print(list(e1))
print(list(e2))
```

```python
# for in 사용
# 리스트를 통해 반복을 하기 때문에 v1 변수에는 리스트가 관리하는 반복 번째의 
# 값이 담기된다.
# 하지만 여기에서 지금 몇 번째 반복인지는 알 수가 없다.
for v1 in list1 :
    print(v1)
```

```python
# range 사용
for v1 in range(len(list1)) :
    # v1에는 0 부터 1씩 증가되는 값이 담기기 때문에
    # 이것을 통해서 몇 번째 반복인지 파악이 가능하다
    print(v1)
    # 하지만 값을 추출하기 위해서 직접 값을 가지고 나와야 한다.
    v2 = list1[v1]
    print(v2)
```

```python
# enumerate 사용
# for v1 in enumerate(list1) :
#     print(v1[0])
#     print(v1[1])

# enumerate를 사용하면 (순서값, 실제값) 형태의 튜플들이 만들어지고 
# 이 튜플들을 관리하는 요소가 만들어진다.
# 이것을 이용해 for문을 운영하고 분해 개념을 이용해 순서값과 실제값을 각각
# 변수에 담아서 사용하면 편리하다.
for idx1, v1 in enumerate(list1) :
    print(idx1)
    print(v1)
```

```python
# 딕셔너리
# 데이터의 이름들만 가져온다.
name1 = dict1.keys()
print(name1)
# 값들만 가져온다.
value1 = dict1.values()
print(value1)
# (이름, 값) 형태로 되어 있는 튜플들을 가지고 있는 요소
items1 = dict1.items()
print(items1)
```

```python
# 이름만 가지고 반복하기
for key in dict1.keys() :
    print(key)

print()

# 값만 가지고 반복하기
for value in dict1.values() :
    print(value)

print()

# 이름과 값을 가지고 반복하기
for key, value in dict1.items() :
    print(key, value)
```

```python
# 리스트와 튜플은 for in 을 사용하면 반복 번째 값이 나오지만
# 딕셔너리는 이름이 나온다.
for n1 in dict1 :
    print(n1)
    print(dict1[n1])
```
