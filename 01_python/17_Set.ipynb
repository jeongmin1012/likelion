### set
- 데이터를 집합 형태로 관리할 수 있는 요소
- 중복된 데이터를 허용하지 않는다.
- 각종 집합에 관련된 연산을 수행 수 있다
- 순서에 대한 개념이 없고 이름으로 데이터를 관리하지 않는다

```python
# set 은 중복된 데이터를 허용하지 않는다.
s1 = {10, 10, 20, 20, 30, 30}
print(s1)
```

```python
# (시험에는 안나옴)
# set을 통해 for in을 운영할 수는 있지만 어떠한 순서대로 값이 나올지는 예측하기가 어렵다.
# 이에 list나 tuple로 변환해서 사용하는 것을 권장한다.
list1 = list(s1)
print(list1)
```
