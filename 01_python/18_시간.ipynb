```python
# 시간을 관리하는 기능을 사용하기 위해...
import time
```

```python
# 현재시간 가져오기
# 가져올 수 있는 값은 유닉스 타임 값
# 1970년 1월 1일 0시 0분 0초 0ms (세계 표준시)를 0으로 설정하고
# 1ms 마다 증가 혹은 감소 시킨 값
# 이 시간 값은 컴퓨터 입장에서의 시간 값
# ms 단위의 시간 값
a1 = time.time()
print(a1)

# ns 단위의 시간 값
a2 = time.time_ns()

print(a2)
```

```python
# 사람입장에서의 시간 값
# 컴퓨터에서 시간 값을 가져온다.
a2 = time.localtime()
print(f'타임 존 : {a2.tm_zone}')
print(f'년 : {a2.tm_year}')
print(f'월 : {a2.tm_mon}')
print(f'현재 월에서 몇 번째 일인지 : {a2.tm_mday}')
print(f'시 : {a2.tm_hour}')
print(f'분 : {a2.tm_min}')
print(f'초 : {a2.tm_sec}')
print(f'이번 주에서 몇 번째 일인지 : {a2.tm_wday}')
print(f'올 해에서 몇 번째 일인지 : {a2.tm_yday}')
```
