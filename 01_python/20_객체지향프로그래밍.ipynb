```python
#  객체의 맴버
# 변수 : 맴버 변수
# 함수 : 맴버 메서드

# 파이썬은 클래스를 통해 생성되는 객체의 맴버 변수를 정의할 수 없다.
# 클래스를 통해 객체를 생성하면 아무것도 가지고 있지 않는 텅 비어있는 객체가 생성된다.
# 이 객체에 변수를 추가해줘야 객체가 변수들을 관리할 수 있다.
class TestClass2 :
    # 생성자 함수
    # 클래스를 통해 객체를 생성하면 무조건 호출되는 함수
    # 파이썬에서는 클래스를 통해 객체를 생ㅅ엉할 경우 텅 비어있는 객체를 생성하고
    # 생성자 함수를 호출한다.
    # 여기에서 객체에 변수를 추가하는 작업을 하게 되면 이 클래스를 가지고 만든
    # 모든 객체를 동일한 변수를 가진 객체가 될 수 있다.

    # self : 클래스안에 정의한 함수들이 가지고 있는 매개변수. 변수의 이름은 self가 아니어도 된다.
    # self에는 함수가 호출이 된 원인이 되는 객체가 들어온다.
    # __init__ 은 객체가 생성되면 자동으로 호출되는 함수이기 때문에 함수 호출의 원인이되는 객체가 존재한다.
    # 이 객체가 self 안으로 들어온다.
    def __init__(self) :
        print('init')
        print(f'self : {self}')
        # self를 통해 객체에 접근하여 변수들을 추가해준다
        self.a1 = 100
        self.a2 = 200

t1 = TestClass2()
print(f't1   : {t1}')
print(t1.a1)
print(t1.a2)
```

```python
# 생성자 함수에 매개변수를 정의할 수 있다.
class TestClass3 :
    # 생성자 함수
    def __init__(self, v1, v2, v3 = 3) :
        self.a1 = v1
        self.a2 = v2
        self.a3 = v3

t1 = TestClass3(10, 20, 30)
print(t1.a1, t1.a2, t1.a3)

t2 = TestClass3(100, 200)
print(t2.a1, t2.a2, t2.a3)
```

```python
# 메서드
# 객체를 통해 호출하는 함수들
class TestClass4 :
    # 생성자
    def __init__(self, a1, a2):
        self.v1 = a1
        self.v2 = a2

    # 메서드
    # 객체를 통해서 호출하는 함수를 의미한다.
    # 메서드는 첫 번째 매개변수에 호출할 때 사용한 객체의 정보가 들어온다.
    # 이를 통해서 메서드를 호출할 때 사용한 객체에 접근할 수 있다.
    def showValue(self) :
        print(self.v1)
        print(self.v2)

    def setValue(self, a1, a2) :
        self.v1 = a1
        self.v2 = a2
```

```python
t1 = TestClass4(10, 20)
t2 = TestClass4(100, 200)

t1.showValue()
t2.showValue()
print()

t1.setValue(50, 60)
t2.setValue(500, 600)

t1.showValue()
t2.showValue()
```
