### 상속
- 클래스가 다른 클래스가 가진 것들을 물려받는 개념
- 상속 시켜 주는 클래스 : 부모클래스(Super Class)
- 상속 받는 클래스 : 자식클래스(Sub Class)
- 자식 클래스는 부모가 가진 것들을 마치 자기 자신한테 있는 것처럼 사용할 수 있다.

```python
# 부모 클래스
class Animal :

    def __init__(self, type) :
        self.animal_type = type
    
    def eat(self) :
        print(f'{self.animal_type}가(이) 사료를 먹습니다')
```

```python
# 자식 클래스
class Dog(Animal) :

    def speaking(self) :
        print('멍멍')

class Cat(Animal) :

    def speaking(self) :
        print('냐옹 냐옹')
```

```python
dog = Dog('강아지')
print(dog.animal_type)
dog.eat()
dog.speaking()

cat = Cat('고양이')
print(cat.animal_type)
cat.eat()
cat.speaking()
```
