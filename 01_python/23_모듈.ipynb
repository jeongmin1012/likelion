### 모듈
- 파이썬 코드를 작성하는 파일의 확장자는 py 이다
- 이 py 파일이 하나의 모듈이 된다.
- 파이선 코드를 작성할 때 다른 py 파일에 작성되어 있는 것을 사용할 수 있다.

```python
# test_module.py 파일의 코드를 실행시킨다.
import test_module1
```

```python
# 이미 import 한 모듈을 또 import 하면 아무런 일도 하지 않는다.
import test_module1
```

```python
# 다른 모듈에 있는 변수, 함수, 클래스를 사용한다.
import test_module2
# 위와 같이 모듈명만 작성하여 import를 할 경우
# 해당 모듈에 내에 있는 요소를 사용할 때는 모듈명을 붙혀줘야 한다.
print(test_module2.test2)
test_module2.test2_function()
t2 = test_module2.TestClass2()
print(t2)
```

```python
# 별칭
# import 모듈명 형태로 작성을 해주면 모듈이 가지고 있는 요소들을 사용할 때 마다
# 앞에 모듈명을 붙혀줘야 한다. (바로 위의 예시)
# 만약 모듈명이 너무 길어서 불편하다면 별칭을 부여해 줄 수 있다.
import test_module3 as m3
print(m3.test3)
m3.test3_function()
t3 = m3.TestClass3()
print(t3)
```

```python
# 별칭을 부여해서 import 한 모듈을 모듈명으로는 사용할 수 없고 
# 별칭으로만 사용할 수 있다 (단, 파이썬 버전에 따라 다를 수 있다)
print(test_module3.test3)
```

```python
# 모듈명을 생략 - 사용하고 싶은 것에 대해서만 지정하기
# from 모듈명 import 요소명
from test_module4 import test4
from test_module4 import test4_function, TestClass4

print(test4)
test4_function()
t4 = TestClass4()
print(t4)
```

```python
# 모듈명을 생략 - 사용하고 싶은 것에 대해서만 지정하기
# from 모듈명 import 요소명
from test_module4 import test4
from test_module4 import test4_function, TestClass4

print(test4)
test4_function()
t4 = TestClass4()
print(t4)

# 보통 import는 상단 부분에 몰아서 작성한다. 해서 모듈명을 생략해서 사용하는 방식을 사용하면
# 나중에 해당 요소가 어떤 모듈에 있는 것인지 바로 파악하기가 어려울 수 있다.

# 같은 프로젝트에서 사용하는 다른 모듈에 동일명의 요소(변수, 함수, 클래스 등)이 있을 경우 문제가 발생할 수 있다.
# 즉 모듈명이 생략되어 있기 때문에 해당 요소가 어느 모듈에 있는 것인지 판단할 수 없기 때문에 오류가 발생한다.
# 만약 모듈명을 생략하는 방식으로 import 된 요소와 이름이 같은 다른 모듈에 있는 요소를 사용할 때는
# 다른 요소들은 모듈명을 생략하는 방식으로 작업하면 안된다.
```

```python
# 모듈에 있는 모든 요소들에 대해 모듈을 생략하고 사용하고자 할 때
# 다른 모듈들이 가진 요소의 이름과 중복될 확률이 너무 높다라는 것을 주의해야 한다.
from test_module5 import *

print(test5)
test5_function()
t5 = TestClass5()
print(t5)
```
