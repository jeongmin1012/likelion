### 패키지

- py 파일들을 가지고 있는 폴더(디렉토리)를 의미한다.
- 패키지안에 있는 모듈을 사용할 때도 import를 해줘야 한다.

```python
# 패키지내의 모듈을 import 한다
import package1.test_module6

# 패키지내의 모듈이 가지고 있는 요소를 사용할 때는 패키지명.모듈명.요소명 으로 해야 한다.
print(package1.test_module6.test6)
package1.test_module6.test6_function()
t6 = package1.test_module6.TestClass6()
print(t6)
```

```python
# 패키명을 생략하고자 한다면..
# 모듈 내부의 요소를 사용할 때는 모듈명은 붙혀줘야 한다.
from package2 import test_module7
print(test_module7.test7)
test_module7.test7_function()
t7 = test_module7.TestClass7()
print(t7)
```

```python
# 패키지명과 모듈명을 생략한다.
from package3.test_module8 import test8, test8_function, TestClass8

print(test8)
test8_function()
t8 = TestClass8()
print(t8)
```

```python
# 패키지내의 모든 모듈들을 사용할 때 패키지명을 생략한다.
# 모듈내에 있는 모든 요소를 사용하고자 할 때는 그냥 * 를 사용해주면 된다.
# 하지만 패키지내에 있는 모든 모듈을 사용하고자 할 때 * 를 사용하면 패키지내의 모든 모듈들을 사용할 수 없다.
# from 패키지명 import * 을 하게 되면 해당 패키지 폴더안에 있는 __init__.py 파일을 찾는다.
# 이 파일안에 명시되어 있는 모듈들을 패키지명 없이 사용할 수 있게 된다.
# 원래는 과거의 파이썬에서는 폴더내에 __init__.py 파일이 없으면 패키지로 인식되지 않았다.
# 파이썬 버전이 업데이트 되면서 그냥 py 파일만 있으면 패키지로 인식이 된다.
# 그래도 * 을 통해 패키지명을 생략하고 모듈을 사용하고 싶은 패키지를 만들어주고자 한다면 __init__.py 파일을 만들고
# 모듈의 이름들을 명시해줘야 한다.

from package4 import *

print(test_module9.test9)
test_module9.test9_function()
t9 = test_module9.TestClass9()
print(t9)

print(test_module10.test10)
test_module10.test10_function()
t10 = test_module10.TestClass10()
print(t10)
```
