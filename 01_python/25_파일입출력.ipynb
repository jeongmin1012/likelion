```python
# 파일 쓰기
# 항상 파일을 먼저 열어야 한다.
# 첫 번째 : 파일의 이름
# 두 번째 : 파일을 여는 목적 (w : 쓰기, a : 추가, r : 읽기)
# w나 a 로 파일을 열면 파일이 없을 경우 파일을 생성한다
f = open('data1.txt', 'w')

# 파일에 쓴다.
f.write('안녕하세요\n')
f.write('반갑습니다')

# 파일을 닫는다
f.close()
```

```python
# 읽기
f = open('data1.txt', 'r')

# 첫 번째 줄 부터 읽어온다.
a1 = f.readlines()

# 파일을 닫아준다
f.close()

print(a1)
```

```python
# with 블럭
# 파일 입출력, 데이터 베이스 사용, 네트워크 등등 어디에 접속해서 데이터를 읽어오는
# 작업을 할 때 사용하는 블럭이다.
# with 내부의 코드가 모두 수행되면 자동으로 파일을 닫아준다.
with open('data2.txt', 'w') as fp :
    fp.write('동해물과\n')
    fp.write('백두산이')
```

```python
# 읽기
with open('data2.txt', 'r') as fp :
    a1 = fp.readlines()

print(a1)
```

```python
# pickle
# 객체를 저장할 수 있는 도구(라이브러리)
# 파이썬은 모든 요소들(정수, 실수, 논리, 함수, 객체, 리스트, 튜플, 딕셔너리 등등등 모든 것)을 객체로 관리한다.
# 따라서 객체를 저장할 수 있다는 것은 그냥 모든 것을 저장할 수 있다는 뜻이된다.
import pickle

class TestClass1 :
    def __init__(self, a1, a2) :
        self.v1 = a1
        self.v2 = a2

    def showData(self) :
        print(self.v1)
        print(self.v2)

t1 = TestClass1(10, 20)
t2 = TestClass1(100, 200)

t3 = [10, 20, 30]
t4 = 10, 20, 30
t5 = {'k1' : 100, 'k2' : 200}

# w : 쓴다는 뜻, b : 여러 형태의 값
with open('data3.txt', 'wb') as fp :
    pickle.dump(100, fp)
    pickle.dump(11.11, fp)
    pickle.dump(True, fp)
    pickle.dump("안녕하세요", fp)
    pickle.dump(t1, fp)
    pickle.dump(t2, fp)
    pickle.dump(t3, fp)
    pickle.dump(t4, fp)
    pickle.dump(t5, fp)
```

```python
# pickle을 이용해 저장한 객체들을 읽어올 때는 저장한 순서대로 읽어와야 한다.
with open('data3.txt', 'rb') as fp :
    z1 = pickle.load(fp)
    z2 = pickle.load(fp)
    z3 = pickle.load(fp)
    z4 = pickle.load(fp)
    z5 = pickle.load(fp)
    z6 = pickle.load(fp)
    z7 = pickle.load(fp)
    z8 = pickle.load(fp)
    z9 = pickle.load(fp)

print(z1)
print(z2)
print(z3)
print(z4)
z5.showData()
z6.showData()
print(z7)
print(z8)
print(z9)
```
