### 전체적인 프로그램의 일반적인 구조
- 데이터를 준비
- 준비된 데이터를 가지고 처리를 한다.
- 처리를 통해 얻은 결과를 출력한다.

### 작업순서
- 화면부터 구성한다. (이 때, 값이 필요하다면 임의의 값 아무거나 정해준다)
- 화면을 구성하기 위해 필요한 데이터들이 무엇이 있는지 확인한다.
- 확인된 데이터를 담을 변수를 선언하고 값은 임의의 값을 넣어준다.
- 변수에 담을 값을 어떻게 구할 수 있는지 전략을 세운다.
- 세운 전력대로 값을 구해 변수에 담아준다.

```python
#1단계

# 학생 데이터를 준비한다

# 준비된 학생 데이터를 통해 처리를 한다.

# 처리된 결과를 통해 출력할 화면을 구성한다.
print('이름 : 홍길동')
print('나이 : 10 살')
print('국어점수 : 70 점')
print('영어점수 : 71 점')
print('수학점수 : 72 점')
print('')
print('이름 : 김길동')
print('나이 : 11 살')
print('국어점수 : 80 점')
print('영어점수 : 81 점')
print('수학점수 : 82 점')
print('')
print('이름 : 최길동')
print('나이 : 12 살')
print('국어점수 : 90 점')
print('영어점수 : 91 점')
print('수학점수 : 92 점')
```
