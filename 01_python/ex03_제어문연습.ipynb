```python
#문제1

# 처음에는 1을 출력한다.
# 그 다음 부터는 3씩 증가시킨 값을 출력한다.
# 1을 포함한 총 20개의 숫자를 출력한다.

num = 1

for cnt in range(20):
    print(cnt+1,':',num)
    num=num+3
```

```python
#문제2)
# 1부터 100까지의 숫자 중 3의 배수를 모두 출력한다.

for i in range(100):
    if (i+1)%3 == 0 :
        print(i+1)
    else:
        continue
```

```python
#문제3)
# 사용자에게 숫자를 입력받는다.
# 1부터 입력받은 숫자까지 정수의 합을 구해 출력한다.

num=int(input('숫자를 입력하세요'))
sum=0

for i in range(num):
    sum=sum+(i+1)
    print("i:",i+1,"sum: ",sum)
print(sum)
```

```python
#문제3_2
num=int(input('숫자를 입력하세요'))
sum=0
cnt = 1

while cnt<=num:
    sum=sum+cnt
    print("i:",cnt,"sum: ",sum)
    cnt=cnt+1
print(sum)
```

```python
# 문제4)
# 1부터 1000까지의 숫자 중 소수를 모두 출력한다.
# 소수 : 1과 자기 자신으로만 나누어 떨어지는 숫자를 의미한다.
num = 2
for num in range(1001):
    cnt=0
    for i in range(num//2):
        if num%(i+1) == 0:
            cnt=cnt+1
        else:
            continue
    if cnt==1:
        print(num)
```

```python
#문제4_2

num=1
while num<=1000:
    cnt=0
    i=1
    while i<=num:
        if num%i == 0:
            cnt=cnt+1
        i=i+1
    if cnt==2:
        print(num)
    num=num+1
```

```python
#문제1
print('문제 1')

num = 1 # 처음에는 1을 출력한다.

for cnt in range(20): # 1을 포함한 총 20개의 숫자를 출력한다.
    print(cnt+1,':',num)
    num=num+3 # 그 다음 부터는 3씩 증가시킨 값을 출력한다.

################################################

#문제 2
print()
print('문제 2')

for i in range(100): # 1부터 100까지의 숫자 
    if (i+1)%3  == 0 : #3의 배수를 모두 출력한다.
        print(i+1)
    else:
        continue

#############################################

#문제 3
print()
print('문제 3')

num=int(input('숫자를 입력하세요')) # 사용자에게 숫자를 입력받는다.
sum=0

for i in range(num): # 1부터 입력받은 숫자까지 정수의 합을 구해 출력한다.
    sum=sum+(i+1)
    print("i:",i+1,"sum: ",sum)
print('최종 합계 :',sum)

##############################################

#문제 4
print()
print('문제 4')

num=1
while num<=1000: #1부터 1000까지
    cnt=0
    i=1
    while i<=num: #1부터 자기자신까지
        if num%i == 0: #약수 갯수 구하기
            cnt=cnt+1
        i=i+1
    if cnt==2: #2개여야 소수
        print(num)
    num=num+1
```
