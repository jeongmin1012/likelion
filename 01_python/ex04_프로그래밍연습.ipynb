```python
# 문제 1
num = random.randint(1,100)#1 에서 부터 100까지의 범위안에서 랜덤값 하나를 추출한다.
print(num)
sum=0

for i in range(num): #1 부터 랜덤값 까지의 총합을 구해 출력한다.
    sum=sum+(i+1)
    print('i:',i+1,'sum :',sum)
print('총 합계 :',sum)
```

```python
import random

# 사용자에게 1 에서부터 100까지의 범위안에서 랜덤값 하나를 추출한다.
number1 = random.randint(1, 100)

# 1부터 랜덤값 까지의 총합을 구한다.
total = 0
number2 = 1

# 현재 숫자가 랜덤값보다 작거나 같으면 반복한다.
while number2 <= number1 :
    # 누적을 한다.
    total = total + number2
    # 현재 숫자를 증가한다.
    number2 = number2 + 1

# 총합을 출력한다.
print('랜덤 값 :', number1)
print('총합 :', total)
```

```python
# 첫 번째 숫자를 추출한다.
number1 = random.randint(1, 45)

# 두 번째 숫자를 추출한다.
while True :
    # 숫자를 추출한다.
    number2 = random.randint(1, 45)
    # 만약 이전에 추출하지 않은 숫자가 추출되었다면 반복을 중단한다.
    if number1 != number2 :
        break


# 세 번째 숫자를 추출한다.
while True :
    # 숫자를 추출한다.
    number3 = random.randint(1, 45)
    # 만약 이전에 추출하지 않은 숫자가 추출되었다면 반복을 중단한다.
    if number1 != number3 and number2 != number3 :
        break

# 네 번째 숫자를 추출한다.
while True :
    # 숫자를 추출한다.
    number4 = random.randint(1, 45)
    # 만약 이전에 추출하지 않은 숫자가 추출되었다면 반복을 중단한다.
    if number1 != number4 and number2 != number4 and number3 != number4 :
        break

# 다섯 번째 숫자를 추출한다.
while True :
    # 숫자를 추출한다.
    number5 = random.randint(1, 45)
    # 만약 이전에 추출하지 않은 숫자가 추출되었다면 반복을 중단한다.
    if number1 != number5 and number2 != number5 and number3 != number5 and number4 != number5 :
        break

# 여섯 번째 숫자를 추출한다.
while True :
    # 숫자를 추출한다.
    number6 = random.randint(1, 45)
    # 만약 이전에 추출하지 않은 숫자가 추출되었다면 반복을 중단한다.
    if number1 != number6 and number2 != number6 and number3 != number6 and number4 != number6 and number5 != number6 :
        break

# 추출한 숫자들을 모두 출력한다. 
print(number1, number2, number3, number4, number5, number6)
```

```python
# 1에서 부터 100까지의 범위안에서 랜덤값 하나를 추출한다.
number1 = random.randint(1, 100)
# 사용자에게 입력을 받는다.
input_number = int(input('1~100사이의 정수를 입력해주세요'))
# 사용자가 랜덤값을 맞출 때 까지 반복한다.    
while input_number != number1 :
    # 사용자가 입력한 숫자가 1 ~ 100까지의 범위가 아니라면
    if input_number < 1 or input_number > 100 :
        print('잘못된 입력입니다')
    # 입력받은 숫자가 랜덤값보다 크면 "큽니다"를 출력한다.
    elif input_number > number1 :
        print('큽니다')
    # 입력받은 숫자가 랜덤값 보다 작으면 "작습니다"를 출력한다.
    elif input_number < number1 :
        print('작습니다')
    # 입력받은 숫자가 랜덤값하고 같으면 반복을 중단한다.
    elif input_number == number1 :
        break

    # 사용자에게 입력을 받는다.
    input_number = int(input('1~100사이의 정수를 입력해주세요'))

# 입력받은 숫자가 랜덤값이랑 같으면 "정답입니다"를 출력한다.
print('정답입니다')
```
