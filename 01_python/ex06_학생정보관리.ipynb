```python
#딕셔너리를 관리하는 리스트
students=[]

#처음에 학생 수를 입력받는다.
stu_num = int(input('학생 수를 입력하세요'))

#입력받은 학생 수만큼 정보를 입력받는다.
for i in range (stu_num):
    name, age, kor, eng, math = input(f'{i+1}번 학생의 이름, 나이, 국어점수, 영어점수, 수학점수를 입력하세요(공백으로 구분)').split()

    stu={
        '이름' : name,
        '나이' : int(age),
        '국어점수' : int(kor),
        '영어점수' : int(eng),
        '수학점수' : int(math)
    }

    students.append(stu)

# 총점 계산
kor_sum=0
eng_sum=0
math_sum=0
fsum=0
for stu in students:
    kor_sum=kor_sum+stu['국어점수']
    eng_sum=eng_sum+stu['영어점수']
    math_sum=math_sum+stu['수학점수']

fsum=kor_sum+eng_sum+math_sum

#평균 계산
kor_avg=int(kor_sum/stu_num)
eng_avg=int(eng_sum/stu_num)
math_avg=int(math_sum/stu_num)
avg=int((kor_sum+eng_sum+math_sum)/stu_num/3)

#차례로 정보 출력
for stu in students:
     print(f'이름 : {stu["이름"]}')
     print(f'나이 : {stu["나이"]}살')
     print(f'국어점수 : {stu["국어점수"]}점')
     print(f'영어점수 : {stu["영어점수"]}점')
     print(f'수학점수 : {stu["수학점수"]}점')
     print()

#총 학생의 수 출력
print(f'총 학생의 수 : {stu_num}명')
print()

#과목별 출력
print(f'국어 총점 : {kor_sum}점')
print(f'영어 총점 : {eng_sum}점')
print(f'수학 총점 : {math_sum}점')
print()
print(f'국어 평균 : {kor_avg}점')
print(f'영어 평균 : {eng_avg}점')
print(f'수학 평균 : {math_avg}점')
print()
print(f'전체 총점 : {fsum}점')
print(f'전체 평균 : {avg}점')

```
