```python
#딕셔너리를 관리하는 리스트
students=[]

#처음에 학생 수를 입력받는다.
stu_num = int(input('학생 수를 입력하세요'))
subjects = ['국어', '영어', '수학']

#입력받은 학생 수만큼 정보를 입력받는다.
def stu_info_input(num):
    name, age, kor, eng, math = input(f'{num+1}번 학생의 이름, 나이, 국어점수, 영어점수, 수학점수를 입력하세요(공백으로 구분)').split()

    stu={
        '이름' : name,
        '나이' : int(age),
        '국어' : int(kor),
        '영어' : int(eng),
        '수학' : int(math)
    }
    return stu

# 총점 계산
def total_score(people, *subject):
    sum = 0
    for stu in people:
        for sub in subject:
            sum = sum + stu[sub]
    return sum

#평균 계산
def avg_score(total, *num):
    if len(num) == 1:
        students = num[0]
        return total / students
    
    elif len(num) == 2:
        students = num[0]
        subjects = num[1]
        return total / (students * subjects)

#차례로 정보 출력
def print_stu_info(people):
    for stu in people:
        print(f'이름 : {stu["이름"]}')
        print(f'나이 : {stu["나이"]}살')
        print(f'국어점수 : {stu["국어"]}점')
        print(f'영어점수 : {stu["영어"]}점')
        print(f'수학점수 : {stu["수학"]}점')
        print()

#부가 출력
def print_results(num,totals,avgs):
    print(f'총 학생의 수 : {num}명\n')
    for sub in totals:
        print(f'{sub} 총점 : {totals[sub]}점')
    print()
    for sub in avgs:
        print(f'{sub} 평균 : {avgs[sub]:.1f}점')
    print()

    print(f'전체 총점 : {totals["전체"]}점')
    print(f'전체 평균 : {avgs["전체"]:.1f}점')

#학생 정보 입력 함수
for idx in range(stu_num):
    student = stu_info_input(idx)
    students.append(student)

#총점 계산 함수
totals={}
for sub in subjects:
    totals[sub] = total_score(students, sub)
totals['전체'] = total_score(students, *subjects)

#평균 계산 함수
avgs = {}
for sub in subjects:
    avgs[sub] = avg_score(totals[sub], stu_num)
avgs['전체'] = avg_score(totals['전체'], stu_num, len(subjects))

#출력 함수
print_stu_info(students)
print_results(stu_num, totals, avgs)
```

```python
# 학생수를 입력받는 함수
def inputStudentCnt() :
    studentCnt = input('학생 수를 입력해주세요')
    return int(studentCnt)

# 학생정보를 입력받는 함수
def inputStudentInfo() :
    # 학생의 정보를 받는다.
    name1 = input("이름을 입력해주세요")
    
    a1 = input('나이를 입력해주세요')
    age1 = int(a1)

    a1 = input('국어 점수를 입력해주세요')
    kor1 = int(a1)

    a1 = input('영어 점수를 입력해주세요')
    eng1 = int(a1)

    a1 = input('수학 점수를 입력해주세요')
    math1 = int(a1)

    student_dict = {
        'name' : name1,
        'age' : age1,
        'kor' : kor1,
        'eng' : eng1,
        'math' : math1
    }

    return student_dict

# 각 과목별 총점을 구하는 함수
def getSubjectTotal(studentList) :
    # 각 과목별 총점을 담을 변수
    korTotal = 0
    engTotal = 0
    mathTotal = 0

    # 학생 수 만큼 반복하면서 총점을 구한다.
    for studentInfo in studentList :
        korTotal = korTotal + studentInfo['kor']
        engTotal = engTotal + studentInfo['eng']
        mathTotal = mathTotal + studentInfo['math']

    return korTotal, engTotal, mathTotal

# 각 과목별 평균을 구하는 함수
def getSubjectAvg(korTotal, engTotal, mathTotal, studentCnt) :
    # 평균을 구한다.
    korAvg = korTotal // studentCnt
    engAvg = engTotal // studentCnt
    mathAvg = mathTotal // studentCnt

    return korAvg, engAvg, mathAvg

# 전체 총점, 평균을 구하는 함수
def getSubjectAll(korTotal, engTotal, mathTotal, studentCnt, subjectCnt) :
    # 전체 총점을 구한다.
    allTotal = korTotal + engTotal + mathTotal
    # 전체 평균을 구한다.
    allAvg = allTotal // studentCnt // subjectCnt

    return allTotal, allAvg

# 학생의 정보를 출력하는 함수
def showStudentInfo(studentList) :
    # 학생의 수 만큼 반복한다.
    for studentInfo in studentList :
        print(f'이름 : {studentInfo['name']}')
        print(f'나이 : {studentInfo['age']}살')
        print(f'국어점수 : {studentInfo['kor']}점')
        print(f'영어점수 : {studentInfo['eng']}점')
        print(f'수학점수 : {studentInfo['math']}점')
        print()

# 추가 정보를 출력하는 함수
def showTotalInfo(student_cnt, kor_total, eng_total, math_total, kor_avg, eng_avg, math_avg, all_total, all_avg) :
    print(f'총 학생의 수 : {student_cnt}명')
    print()

    print(f'국어 총점 : {kor_total}점')
    print(f'영어 총점 : {eng_total}점')
    print(f'수학 총점 : {math_total}점')
    print()

    print(f'국어 평균 : {kor_avg}점')
    print(f'영어 평균 : {eng_avg}점')
    print(f'수학 평균 : {math_avg}점')
    print()

    print(f'전체 총점 : {all_total}점')
    print(f'전체 평균 : {all_avg}점')

# 학생수를 입력받는다.
studentCnt = inputStudentCnt()

# 학생 정보를 담을 리스트
studentList = []

# 학생의 수 만큼 반복한다.
for idx1 in range(studentCnt) :
    # 학생 정보를 입력받는다.
    studentInfo = inputStudentInfo()
    studentList.append(studentInfo)

# 과목별 총점을 구한다.
kor_total, eng_total, math_total = getSubjectTotal(studentList)

# 과목별 평균을 구한다.
kor_avg, eng_avg, math_avg = getSubjectAvg(kor_total, eng_total, math_total, studentCnt)

# 전체 총점, 편균을 구한다.
all_total, all_avg = getSubjectAll(kor_total, eng_total, math_total, studentCnt, 3)

# 학생 정보를 출력한다.
showStudentInfo(studentList)

# 추가 정보를 출력한다.
showTotalInfo(studentCnt, kor_total, eng_total, math_total, kor_avg, eng_avg, math_avg, all_total, all_avg)
```
