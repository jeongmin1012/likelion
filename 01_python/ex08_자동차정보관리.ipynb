```python
#자동차 객체
class Car :
    def __init__ (self,name, age,tire,fuel,cc):
        self.name = name
        self.age = age
        self.tire = tire
        self.fuel = fuel
        self.cc = cc

#차고 객체
class Garage :
    def __init__(self):
        self.cars=[]
    
    #자동차 갯수 입력
    def CntCar(self):
        Cnt=int(input('총 몇대의 자동차를 등록하시겠습니까?'))
        return Cnt
    
    #새로운 자동차 등록
    def NewCar(self,Cnt):
        for idx in range(Cnt):
            name = input(f'{idx+1}번 차량의 차종을 입력해주세요.')
            age = int(input(f'{idx+1}번 차량의 연식을 입력해주세요.'))
            tire = int(input(f'{idx+1}번 차량의 타이어 갯수를 입력해주세요.'))
            fuel = int(input(f'{idx+1}번 차량의 연비를 입력해주세요.'))
            cc = int(input(f'{idx+1}번 차량의 배기량을 입력해주세요.'))
            
            car = Car(name,age,tire,fuel,cc)
            self.cars.append(car)
            
            print(f'{idx+1}번 차량의 등록이 완료되었습니다.')
            print()
    
    #합계 연산
    def CarTotal(self):
        totals = {'tire': 0, 'fuel': 0, 'cc': 0}
        for car in self.cars:
            totals['tire']+= car.tire
            totals['fuel']+= car.fuel
            totals['cc']+= car.cc
        return totals
    
    #평균 연산
    def CarAvg(self,totals,Cnt):
        Avgs={'tire': 0, 'fuel': 0, 'cc': 0}
        Avgs['tire']=totals['tire']/Cnt
        Avgs['fuel']=totals['fuel']/Cnt
        Avgs['cc']=totals['cc']/Cnt
        return Avgs
    
    #자동차 정보 출력
    def PrintCar(self):
        for car in self.cars:
            print(f"차종 : {car.name}")
            print(f"연식 : {car.age}년")
            print(f"타이어 갯수 : {car.tire}개")
            print(f"연비 : {car.fuel}km/L")
            print(f"배기량 : {car.cc}cc")
            print()
    
    #부가 정보 출력
    def PrintAll(self,Cnt):
        print(f'자동차 대수 : {Cnt}')
        print()
        totals = self.CarTotal()
        avgs = self.CarAvg(totals, Cnt)
        print(f'타이어 개수 총합 : {totals['tire']}개')
        print(f'연비 총합 : {totals['fuel']}km/L')
        print(f'배기량 총합 : {totals['cc']}cc')
        print()
        print(f'타이어 개수 평균 : {avgs['tire']}개')
        print(f'연비 평균 : {avgs['fuel']}km/L')
        print(f'배기량 평균 : {avgs['cc']}cc')


g = Garage()  
count = g.CntCar()
g.NewCar(count)
totals = g.CarTotal()
avgs = g.CarAvg(totals, count)
g.PrintCar()
g.PrintAll(count)
```

```python
# 작업의 순서
# 1. 각 클래스에 구현해야할 기능들(메서드 들)을 정리한다.
# 2. 각 기능들(메서드 들)에서 필요한 데이터들을 정리한다.
# 3. 각 메서드들을 구현한다.

# 자동차 클래스
class CarClass : 

    def __init__(self) :
        # 이름
        self.name = ''
        # 나이
        self.age = 0
        # 타이어개수
        self.tireCount = 0
        # 연비
        self.fuelEconomy = 0
        # 배기량
        self.dspl = 0

    # 자동차의 정보를 입력받는다.
    def inputCarInfo(self) :
        self.name = input('이름을 입력해주세요')

        a1 = input('나이를 입력해주세요')
        self.age = int(a1)

        a1 = input('타이어 개수를 입력해주세요')
        self.tireCount = int(a1)

        a1 = input('연비를 입력해주세요')
        self.fuelEconomy = int(a1)

        a1 = input('배기량을 입력해주세요')
        self.dspl = int(a1)

    # 자동차의 정보를 출력한다.
    def showCarInfo(self) :
        print(f'이름 : {self.name}')
        print(f'나이 : {self.age}')
        print(f'타이어 개수 : {self.tireCount}개')
        print(f'연비 : {self.fuelEconomy}km/L')
        print(f'배기량 : {self.dspl}cc')
        print()

# 차고 클래스
class GarageClass :

    def __init__(self) :
        # 자동차 대수
        self.carCount = 0
        # 타이어 개수 총합
        self.totalTireCount = 0
        # 연비 총합
        self.totalFuelEconomy = 0
        # 배기량 총합
        self.totalDspl = 0
        # 타이어 개수 평균
        self.avgTireCount = 0
        # 연비 평균
        self.avgFuelEconomy = 0
        # 배기량 평균
        self.avgaDspl = 0
        # 자동차 객체를 담을 리스트
        self.carList = []


    # 자동차의 대수입력받는다.
    def inputCarCount(self) :
        a1 = input('자동차 대수를 입력해주세요')
        self.carCount = int(a1)

    # 부가적인 정보를 출력한다.
    def showTotalInfo(self) :
        print(f'자동차 대수 : {self.carCount}대')
        print()

        print(f'타어어 개수 총합 : {self.totalTireCount}개')
        print(f'연비 총합 : {self.totalFuelEconomy}km/L')
        print(f'배기량 총합 : {self.totalDspl}cc')
        print()

        print(f'타이어 개수 평균 : {self.avgTireCount}개')
        print(f'연비 평균 : {self.avgFuelEconomy}km/L')
        print(f'배기량 평균 : {self.avgaDspl}cc')

    # 부가적인 정보를 구한다.
    def getTotalInfo(self) :
        # 각 부가적인 정보들의 총합을 구한다.
        for car in self.carList :
            self.totalTireCount = self.totalTireCount + car.tireCount
            self.totalFuelEconomy = self.totalFuelEconomy + car.fuelEconomy
            self.totalDspl = self.totalDspl + car.dspl

        self.avgTireCount = self.totalTireCount // self.carCount
        self.avgFuelEconomy = self.totalFuelEconomy // self.carCount
        self.avgaDspl = self.totalDspl // self.carCount

    # 자동차들의 정보를 입력받는다.
    def inputCarsInfo(self) :
        # 자동들의 수 만큼 반복한다.
        for _ in range(self.carCount) :
            # 자동차의 객체를 생성한다.
            car = CarClass()
            # 자동차의 정보를 입력받는다.
            car.inputCarInfo()
            # 리스트에 담는다.
            self.carList.append(car)

    # 자동차들의 정보를 출력한다.
    def showCarsInfo(self) :
        # 자동차들의 수 만큼 반복한다.
        for car in self.carList :
            car.showCarInfo()
```

```python
# 차고 객체를 생성한다.
garage = GarageClass()
# 자동차의 대수를 입력받는다.
garage.inputCarCount()
# 자동차들의 정보를 입력받는다.
garage.inputCarsInfo()
# 자동차들의 부가 정보를 구한다.
garage.getTotalInfo()
# 자동차들의 정보를 출력한다.
garage.showCarsInfo()
# 자동차들의 부가 정보를 출력한다.
garage.showTotalInfo()
```
