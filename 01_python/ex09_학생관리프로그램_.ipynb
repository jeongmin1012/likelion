```python
# 학생클래스
class StudentClass :
    # 학생 한명의 정보를 입력받는다.
    def inputStudentInfo(self) :
        pass

    # 학생 정보를 출력한다.
    def showStudentInfo(self) :
        pass

# 학교클래스
class SchoolClass :
    # 메뉴를 출력하고 메뉴 번호를 입력받는다.
    def inputMenuNumber(self) :
        pass

    # 학생 한명의 정보를 입력받는다.
    def inputStudentInfo(self) :
        pass

    # 학생들의 정보를 출력한다.
    def showStudentInfo(self) :
        pass

    # 통계 정보를 구하고 출력한다.
    def showStudentAllInfo(self) :
        pass

    # 학생의 정보를 삭제한다.
    def deleteStudentInfo(self) :
        pass


    # 프로그램을 종료한다.
    def exitProgram(self) :
        pass
```

```python
import os
import pickle

# 학생클래스
class StudentClass :

    def __init__(self) :
        self.name = ''
        self.age = 0
        self.kor = 0
        self.eng = 0
        self.math = 0

    # 학생 한명의 정보를 입력받는다.
    def inputStudentInfo(self) :
        self.name = input('이름을 입력해주세요')

        a1 = input('나이를 입력해주세요')
        self.age = int(a1)

        a1 = input('국어점수를 입력해주세요')
        self.kor = int(a1)

        a1 = input('영어점수를 입력해주세요')
        self.eng = int(a1)

        a1 = input('수학점수를 입력해주세요')
        self.math = int(a1)

    # 학생 정보를 출력한다.
    def showStudentInfo(self) :
        print(f'이름 : {self.name}')
        print(f'나이 : {self.age}')
        print(f'국어점수 : {self.kor}')
        print(f'영어점수 : {self.eng}')
        print(f'수학점수 : {self.math}')
        print()

# 학교클래스
class SchoolClass :

    def __init__(self) :
        # 학생 객체를 담을 리스트
        self.studentList = []

        # 저장된 학생 데이터가 있다면 읽어온다.
        if os.path.isfile('student.dat') :
            with open('student.dat', 'rb') as fp :
                self.studentList = pickle.load(fp)


    # 메뉴를 출력하고 메뉴 번호를 입력받는다.
    def inputMenuNumber(self) :
        print('1. 학생정보추가')
        print('2. 학생정보출력')
        print('3. 통계정보출력')
        print('4. 학생정보삭제')
        print('5. 종료')
        print()

        a1 = input('메뉴를 선택해주세요')
        return int(a1)

    # 학생 한명의 정보를 입력받는다.
    def inputStudentInfo(self) :
        # 학생 객체를 생성한다.
        student = StudentClass()
        # 학생의 정보를 입력받는다.
        student.inputStudentInfo()
        # 학생 객체를 리스트에 담아준다.
        self.studentList.append(student)
        # 파일에 리스트를 저장한다.
        with open('student.dat', 'wb') as fp :
            pickle.dump(self.studentList, fp)

    # 학생들의 정보를 출력한다.
    def showStudentInfo(self) :
        # 학생의 수 만큼 반복한다.
        for student in self.studentList :
            student.showStudentInfo()

    # 통계 정보를 구하고 출력한다.
    def showStudentAllInfo(self) :
        # 학생수
        studentCount = len(self.studentList)
        # 학생의 수 만큼 반복한다.
        totalKor = 0
        totalEng = 0
        totalMath = 0

        for student in self.studentList :
            totalKor = totalKor + student.kor
            totalEng = totalEng + student.eng
            totalMath = totalMath + student.math

        avgKor = totalKor / studentCount
        avgEng = totalEng / studentCount
        avgMath = totalMath / studentCount

        print(f'학생수 : {studentCount}')
        print(f'국어총점 : {totalKor}')
        print(f'영어총점 : {totalEng}')
        print(f'수학총점 : {totalMath}')
        print(f'국어평균 : {avgKor}')
        print(f'영어평균 : {avgEng}')
        print(f'수학평균 : {avgMath}')

    # 학생의 정보를 삭제한다.
    def deleteStudentInfo(self) :
        a1 = input('삭제할 학생 순서값')
        self.studentList.pop(int(a1))

        # 파일에 리스트를 저장한다.
        with open('student.dat', 'wb') as fp :
            pickle.dump(self.studentList, fp)


    # 프로그램을 종료한다.
    def exitProgram(self) :
        print('프로그램을 종료합니다')
        exit(0)
```

```python
school = SchoolClass()

while True :
    # 메뉴를 출력한다.
    menuNumber = school.inputMenuNumber()

    if menuNumber == 1 :
        school.inputStudentInfo()
    elif menuNumber == 2 :
        school.showStudentInfo()
    elif menuNumber == 3 :
        school.showStudentAllInfo()
    elif menuNumber == 4 :
        school.deleteStudentInfo()
    elif menuNumber == 5 :
        print('프로그램을 종료합니다')
        break
```
