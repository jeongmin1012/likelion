```python
#동물관리 프로그램

#동물관리 리스트
animals=[]

#처음에는 동물의 수를 입력받는다.
ani_cnt = int(input('동물은 몇마리인가요?(숫자만)'))

#동물의 수 만큼 동물의 정보를 입력받는다.
for cnt in range(ani_cnt):
    #동물의 정보는 이름, 나이, 다리 개수, 먹는 양, 자는 시간,(기타 특이사항)
    name, age, leg, eat, sleep, extra = input(f'{cnt+1}번 동물의 이름, 나이, 다리 개수, 먹는 양, 자는 시간, 기타 특이사항을 입력하세요(공백으로 구분)').split()

    ani={
        'name' : name,
        'age' : int(age),
        'leg' : int(leg),
        'eat' : int(eat),
        'sleep' : int(sleep),
        'extra' : extra
    }
    animals.append(ani)
#나이의 총합
total_age = 0
total_leg = 0
total_eat = 0
total_sleep = 0
for ani in animals:
    total_age= total_age + ani['age']
    total_leg = total_leg + ani['leg']
    total_eat = total_eat + ani['eat']
    total_sleep = total_sleep + ani['sleep']



#모든 동물의 정보를 입력받으면 입력받은 동물의 정보를 출력한다.
for cnt,ani in enumerate(animals):
    print(f'{cnt+1}번째 동물 정보')
    print(f'이름 : {ani['name']}')
    print(f'나이 : {ani['age']}살')
    print(f'다리 개수 : {ani['leg']}개')
    print(f'먹는 양 : {ani['eat']}g')
    print(f'자는 시간 : {ani['sleep']}시간')
    print(f'기타 특이사항 : {ani['extra']}')
    print()
#동물의 수, 나이의 총합, 나이의 평균, 다리의 총 개수, 먹는 총 양, 자는 시간 평균도 출력한다.
print('---------------------')
print(f'동물의 수 : {ani_cnt}마리')
print()
print(f'나이의 총합 : {total_age}살')
print(f'나이의 평균 : {total_age/ani_cnt}살')
print(f'다리의 총 개수 : {total_leg}개')
print(f'먹는 총 양 : {total_eat}g')
print(f'자는 시간 평균 : {total_sleep/ani_cnt}시간')
```
