### Queue
- 스택은 제일 마지막에 넣어준 값이 제일 처음 나오게 된다(Last In First Out, LIFO)
- 큐는 제일 처음에 넣어준 값이 제일 처음 나오게 된다(First In First Out, FIFO)
- 값을 넣어주는 것을 enqueue, 값을 추출하는 것을 dequeue 라고 부른다.

### enqueue
- 새로운 노드를 생성한다.
- 만약 front가 None이면 새로운 노드를 넣어준다.
- 만약 rear가 None이 아니면 rear가 가르키는 
  노드의 다음 노드를 새로운 노드로 설정해준다.
- rear에 새롭게 만든 노드를 넣어준다.

### dequeue
- front 가 None이면 중단한다.
- front가 가르키는 노드의 값을 반환한다.
- front가 가르키는 노드의 다음노드를 front에 넣어준다.
- 제거되는 노드의 다음노드에 None을 넣어준다.
- 만약 제일 마지막 노드를 제거했다면 front와 rear 모두 None을 넣어준다.

```python
class DataNode :
    
    def __init__(self, _value) :
        # 관리할 값
        self.value = _value
        self.nextNode = None
```

```python
class LikeLionQueue :
    
    def __init__(self) :
        # 제일 위,아래
        self.headNode = None
        self.tailNode = None
    
    def enqueue(self,_value):
        newNode = DataNode(_value)
        if self.headNode == None:
            self.headNode = newNode
        
        if self.tailNode != None:
            self.tailNode.nextNode = newNode
        
        self.tailNode = newNode
    
    def dequeue(self):

        if self.headNode == None:
            return None
        value = self.headNode.value
        self.headNode = self.headNode.nextNode

        return value
```

```python
queue = LikeLionQueue()

queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

a1 = queue.dequeue()
print(a1)
a1 = queue.dequeue()
print(a1)
a1 = queue.dequeue()
print(a1)
a1 = queue.dequeue()
print(a1)
```

### Teacher's Version

```python
class DataNode :
    
    def __init__(self, _value) :
        # 관리할 값
        self.value = _value
        # 다음 노드
        self.nextNode = None
```

```python
class LikeLionQueue :
    
    def __init__(self) :
        # font pointer
        self.frontPointer = None
        # rear pointer
        self.rearPointer = None

    # 데이터 추가하기
    def enqueue(self, _value) :
        # 새로운 노드를 생성한다.
        newNode = DataNode(_value)
        # 만약 front가 None이면 새로운 노드를 넣어준다.
        if self.frontPointer == None :
            self.frontPointer = newNode
        # 만약 rear가 None이 아니면 rear가 가르키는 노드의 다음 노드를 새로운 노드로 설정해준다.
        if self.rearPointer != None :
            self.rearPointer.nextNode = newNode
        # rear에 새롭게 만든 노드를 넣어준다.
        self.rearPointer = newNode

    # 데이터 가져오기
    def dequeue(self) :
        # front 가 None이면 중단한다.
        if self.frontPointer == None : 
            print('큐는 비어있습니다')
            return
        
        # front가 가르키는 노드의 값을 반환한다.
        returnValue = self.frontPointer.value
        # front가 가르키는 노드를 변수에 담아준다.
        deleteNode = self.frontPointer
        # front가 가르키는 노드의 다음노드를 front에 넣어준다.
        self.frontPointer = deleteNode.nextNode
        # 제거되는 노드의 다음노드에 None을 넣어준다.
        deleteNode.nextNode = None
        # 만약 제일 마지막 노드를 제거했다면 front와 rear 모두 None을 넣어준다.
        if self.frontPointer == None :
            self.rearPointer = None
        
        return returnValue
```

```python
likelionQueue = LikeLionQueue()

likelionQueue.enqueue(10)
likelionQueue.enqueue(20)
likelionQueue.enqueue(30)

a1 = likelionQueue.dequeue()
a2 = likelionQueue.dequeue()
a3 = likelionQueue.dequeue()

print(a1)
print(a2)
print(a3)
```
