```python
class Stack :
    def __init__(self) :
        self.dataList = []

    def push(self, _value) :
        self.dataList.append(_value)

    def pop(self) :
        a1 = self.dataList.pop()
        return a1
    
class Queue :
    def __init__(self) :
        self.dataList = []

    def enqueue(self, _value) :
        self.dataList.append(_value)

    def dequeue(self) :
        a1 = self.dataList.pop(0)
        return a1
```

```python
class Palindrome:
    def __init__(self):
        self.text=''
    
    def check(self,_test):
        self.text = _test.upper()
        print('upper:',self.text)
        stack =[]
        queue =[]

        for i in self.text:
            stack.append(i)
            queue.append(i)
        
        while stack:
            if stack.pop() != queue.pop(0):
                return False
        
        return True

test = '코난못난코'

palindrome = Palindrome()

print('test: ',test)
print(palindrome.check(test))
```

```python
# 회문
# 회문이란 글자수가 단어나 문장에서 중앙을 기준으로 좌측과 우측이 반전형태로 되어 있는
# 문자열을 의미한다.
# 예) 기러기, 토마토, 스위스, 인도인, 별동별, 우영우

# 매개변수로 홀수인 문자열을 입력받아 회문인지 확인하는 함수를 만들어준다.
# 회문이면 True를 반환하고 아니면 False를 반환한다.
# 함수의 이름은 palindrome 으로 한다

# 확인할 문자열은 대문자로 변경하고 확인한다.

# 힌트) 스택과 큐를 사용하시면 편해집니다....

def palindrome(text) :
    text = text.upper()
    # 문자열을 구성하는 단어들을 추출하여 stack과 queue 에 담는다.
    # stack
    stack = list(text)
    # queue
    queue = list(text)

    # stack이나 queue 글자가 남아있는 동안 반복한다.
    while len(stack) > 0 :
        # stack과 queue에 글자 하나씩을 추출하여 비교한다.
        a1 = stack.pop()
        a2 = queue.pop(0)

        if a1 != a2 :
            # 다른 글자가 발견되면 회문이 아니므로 False을 반환하고 중단한다.
            return False

    # 다른 글자를 발견하지 못하면 True를 반환한다.
    return True

r1 = palindrome('이순신')
print(r1)
```

```python
# 문장에 들어있는 괄호가 올바로 되어 있는지를 확인하는 함수를 만들어준다.
# 괄호는 ( ), { }, < >, [ ] 로 정한다.
# <b>동해물과 {백두(산)이} <[마르]고 {닳도}>록 : True
# <(b>동해물과 {백두(산)이} <[마르]고 {닳도}>록 : False

# 함수의 이름은 check_brackets 라고 한다.
# 괄호가 올바르게 구성되어 있으면 True를 반환하고
# 아니면 False를 반환한다.

# 힌트. 스택을 이용해보세요

def check_brackets(text) :
    # 괄호를 담을 스택을 만든다.
    bracketStack = []

    # 문자열의 처음부터 끝까지 반복한다.
    for c1 in text :
        #  만약 여는 괄호라면 스택에 담는다.
        if c1 in ['(', '<', '[', '{'] :
            bracketStack.append(c1)
        # 만약 여는 괄호라면 스택에서 추출한다.
        if c1 in [')', '>', ']', '}'] :
            a1 =  bracketStack.pop()
            # 만약 추출된 괄호가 짝이 맞지 않으면 옳바르지 않다고 판단한다.
            if a1 == '(' and c1 != ')' :
                return False
            elif a1 == '<' and c1 != '>' :
                return  False
            elif a1 == '[' and c1 != ']' :
                return False
            elif a1 == '{' and c1 != '}' :
                return False

    # 만약 남는 괄호가 없다면 True를 반환한다.
    if len(bracketStack) == 0 :
        return True
    else :
        return False

result = check_brackets('<b>동해물과 {백두(산)이} <[마르]고 {닳도}>')
print(result)
```

```python
def reverse(text):
    save = []
    rever=[]

    for i in text:
        save.append(i)
    
    while len(save)!=0:
        rever.append(save.pop())
    return rever

print(reverse('abcdefg'))
```

```python
# 데이터를 담을 노드
class DataNode :

    def __init__(self, _value) :
        self.value = _value
        self.nextNode = None

    # 다음 노드를 연결하는 함수
    def setNextNode(self, _nextNode) :
        self.nextNode = _nextNode


class Week:
    def __init__(self) :
        # HeadNode (더미 노드)
        self.headNode = DataNode(None)
        self.dataCount = 0
    
    # 요일 데이터를 연결 리스트에 추가
    def addDays(self, days):
        current = self.headNode  

        # 마지막 노드까지 이동해두기
        while current.nextNode is not None:
            current = current.nextNode

        # 리스트에 담긴 요일들을 순서대로 연결
        for d in days:
            newNode = DataNode(d)
            current.setNextNode(newNode)
            current = newNode
            self.dataCount += 1

    # 연결 리스트 출력
    def printAll(self):
        current = self.headNode.nextNode  # 첫 실제 노드

        while current is not None:
            print("value:", current.value)
            current = current.nextNode
        
        print("final node")


# 실제 실행
week = Week()
week.addDays('일월화수목금토')
week.printAll()
```
