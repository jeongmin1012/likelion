### 선형 검색 알고리즘

- 찾고자 하는 데이터를 앞에서 부터 순서대로 이동하면서 찾는 알고리즘

### 예시 그림

```python
# _data_list : 값들이 들어있는 리스트
# _value : 찾을 값
def sequential_search(_data_list, _value) :
    # 리스트의 처음부터 반복한다
    for position, v1 in enumerate(_data_list) :
        # 반복 번째의 값이 _value와 같으면 값의 위치값을 반환한다.
        if v1 == _value :
            return position

    # 만약 찾지 못했다면 None을 반환한다
    return None
```

```python
a1 = [40, 20, 10, 50, 30, 70]

r1 = sequential_search(a1, 20)
r2 = sequential_search(a1, 80)

print(r1)
print(r2)
```
