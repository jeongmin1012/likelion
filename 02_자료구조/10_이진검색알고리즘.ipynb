### 이진 검색 알고리즘
- 데이터가 많을 경우에 사용하는 검색 알고리즘
- 데이터의 위치와 관계 없이 소요되는 작업의 횟수가 비슷하다.
- 데이터가 정렬되어 있어야만 가능한 데이터 검색 알고리즘이다.

```python
# _data_list : 정렬된 데이터가 들어있는 리스트
# _value : 찾고자 하는 값
def binary_search(_data_list, _value):
    # 최소 위치값
    min_position = 0
    # 최대 위치값
    max_position = len(_data_list) - 1

    # 만약 찾고자 하는 값이 없다면 None을 반환한다.
    if _value not in _data_list :
        return None

    # 반복
    while True :
        # 중간 위치값을 계산한다.
        median_position = (max_position + min_position) // 2
        # 중간 위치의 값과 찾고자 하는 값이 같다면..
        if _data_list[median_position] == _value :
            # 찾은 위치값을 반환한다.
            print('찾았습니다')
            return median_position
        # 중간 위치의 값이 찾고자 하는 값보다 크다면..
        elif _data_list[median_position] > _value :
            print('중앙값이 큽니다')
            # 순서의 최대값을 중간 위치값 -1로 설정한다.
            max_position = median_position - 1
        # 중간 위치의 값이 찾고자 하는 값보다 작다면..
        elif _data_list[median_position] < _value :
            print('중앙값이 작습니다')
            # 순서 최소값을 중간 위치값 + 1로 설정한다.
            min_position = median_position + 1
```

```python
# 위의 검색 알고리즘은 데이터가 오름차순 정렬 되어 있다는 것을 가정하였다.
a1 = list(range(100))

p1 = binary_search(a1, 99)
print(p1)
```

```python
# list에 있는 index 함수는 2진탐색 알고리즘을 사용한다.
# 만약 데이터가 정렬되어 있지 않으면 데이터를 정렬을 한다.
# 이 때, 각 데이터의 순서값을 따로 담은 다음에 이 순서값도 같이 정렬한다.
# 이진 탐색을 통해 값을 찾은다음 찾은 값의 순서값을 순서값이 담긴 리스트에서 추출하여 반환한다.
p1 = a1.index(27)
print(p1)
```
