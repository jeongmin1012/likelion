### 삽입정렬
- 정렬이 되어 있는 부분과 정렬이 되어 있지 않은 부분으로 나눈다(처음에는 첫 데이터만 정렬된 부분으로 취급한다)
- 정렬이 안되어 있는 부분의 첫 번째 값을 정렬이 되어 있는 부분 뒤쪽 값부터 하나씩 비교를 한다.
- 정렬이 안되어 있는 부분의 첫 번째 값보다 작은 값을 발견 하면 작은 값 바로 뒤에 삽입한다.

```python
def insert_sort(_data_list):
    final = []

    for i in _data_list:
        if len(final) == 0:
            final.append(i)
        else:
            inserted = False
            for j in range(len(final)-1, -1, -1):
                if final[j] <= i:
                    final.insert(j+1, i)
                    inserted = True
                    break

            if not inserted:
                final.insert(0, i)

    for i in range(len(final)):
        _data_list[i] = final[i] 

    return _data_list
```

```python
# 삽입 정렬 함수
def insert_sort(_data_list) :
    # 전체 데이터의 수 - 1 만큼 반복한다.
    for idx in range(len(_data_list) - 1) :
        print(_data_list)
        # 정렬이 되어 있는 쪽 마지막 순서값
        sorted_last_index = idx
        # 정렬이 되어 있지 않은 쪽 첫 번째 순서값
        unsorted_first_index = idx + 1
        # 반복한다.
        while True :
            # sorted_last_index 번째 값(A)과 unsorted_first_index 번째 값(B)을 비교한다.
            # 만약 A 가 B 보다 작으면 B를 pop 하고 A의 위치 + 1 번째에 B를 삽입한다 (중단)
            if _data_list[sorted_last_index] < _data_list[unsorted_first_index] :
                # unsorted_first_index 번째 값을 pop 한다.
                temp = _data_list.pop(unsorted_first_index)
                # sorted_last_index + 1 번째 삽입한다.
                _data_list.insert(sorted_last_index + 1, temp)
                break
            else :
            # 만약 그렇지 않다면 sorted_last_index를 하나 감소 시킨다.
                sorted_last_index = sorted_last_index - 1
                # 만약 감소된 sorted_last_index가 -1이 되면 B를 pop 하고 0 번째 위치에 삽입한다 (중단)
                if sorted_last_index < 0 :
                    # unsorted_first_index 번째 값을 pop 한다.
                    temp = _data_list.pop(unsorted_first_index)
                    _data_list.insert(0, temp)
                    break
```

```python
a1 = [9, 8, 2, 4, 7, 6, 5, 3, 1]
insert_sort(a1)
print(a1)
```
