### numpy
- 파이썬에서 행렬 데이터를 수학적으로 관리하고 다루기 위해 사용하는 라이브러리
- 일반적인 프로그래밍에서 데이터를 다루는 방식이 아닌 수학에서 행렬을 다루듯이 데이터를 관리하는 라이브러리
- 데이터 분석, AI 와 관련된 모든 라이브러리라 내부적으로 numpy를 사용하고 있다.
- 수학에서는 선형대수학, 공간 벡터 등등 고차원 행렬 계산을 쉽게 할 수 있도록 제공되고 있다.

### scipy
- 과학 산술 연산을 위해 만들어진 라이브러리
- 물리학 등에서 사용하는 연산에 관련된 기능들이 제공된다.
- 딥러닝 알고르짐을 구현할 때 주로 사용한다.
- 개념도 어렵고 사용하기가 어렵다.

### sympy
- 미분, 적분, 벡터, 공간 등의 계산을 위해 만들어진 라이브러리
- 딥러닝 알고리즘을 구현할 때 주로 사용한다.
- 개념도 어렵고 사용하기도 어렵다.
- 미적분을 제외한 나머지 부분들은 numpy에서 제공하고 있다.

### pandas
- 데이터 분석가에게 가장 중요한 라이브러리이다.
- numpy는 수학의 행렬 데이터를 다루듯이 데이터를 관리하지만 pandas는 데이터 베이스의 테이블의 개념과 유하게 데이터를 다룬다.
- 데이터 분석, 시각화, 전처리, 변환, 가공 등등 많은 기능을 제공하고 있다.

### scikit-learn
- 파이썬에서 사용하고 있는 대표적인 머신러닝 알고리즘 라이브러리
- 딥러닝은 없다

### tensorflow, pytorch
- 파이썬에서 사용하고 있는 대표 딥러닝 알고리즘 라이브러리

### matplotlib
- 파이썬에서 사용되어지고 있는 대표 시각화 라이브러리이다.
- 제공되는 그래프의 종류는 많지는 않지만 좌표 평면에 자유롭게 그림을 그릴 수 있다.
- 파이썬에서 사용하고 있는 수 많은 데이터 시각화 라이브러리가 matplotlib를 통해 만들어졌다.

### seaborn
- 파이썬에서 사용되어지고 있는 대표 시각화 라이브러리이다.
- 통계를 기반으로 시각화할 때 주로 사용한다.

### plotly
- 강사 추천

### bs4, selenium
- 웹에서 데이터를 수집하기 위해 사용하는 라이브러리이다.
