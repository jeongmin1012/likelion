### 데이터의 종류
- 범위형 데이터, 수치 데이터(연속성을 가진) : 연속성을 가지는 데이터, 최소나 최대가 존재할 수 도 있다 (나이, 몸무게, 키, 온도, 시간 등등)
- 범주형 데이터, 카테고리 데이터 : 연속성을 가지지 않는 독립적인 데이터(성별, 혈액형 등등)
- 시계열 데이터 : 범위형 데이터의 한 종류이다. 시간의 흐름에 따라 변화되는 데이터를 의미한다(시간에 따른 온도 변화, 시간에 따른 주식 시세)

### 잘못된 데이터의 종류
- 결측치 : 측정되지 않은 데이터. 비어있는 공간을 의미
- 파이썬에서는 NaN(숫자, Not a Number), NaT(시간, Not a Time), Na(그 외, Not  Application, Not Available)로 되어 있다. 현재는 문자열 컬럼도 NaN으로 표시를 한다.
- 이상치 : 잘못된 데이터. 사회 통념상 문제가 있거나 극단치에 해당하는 데이터들은 이상치일 가능성이 있다. 이상치 여부의 정확한 판단은 데이터 명세서를 기반으로 해야 한다.

### 잘못된 데이터의 처리 방법
- 제거 : 추천. 데이터가 왜곡되지 않지만 데이터의 양이 줄어든다.
- 대체 : 데이터의 양이 줄어들지는 않지만 데이터가 왜곡된다. 다중 공선성의 문제가 발생할 확률이 매우 높아진다.

### pandas 사용시 볼 수 있는 특수한 값들
- NaN : Not a Number, 결측치를 의미한다.
- NaT : Not a Time, 결측치를 의미한다.
- Inf : 무한대라는 의미를 가지고 있으며 음수 무한대일 수도 있고 양수 무한대일 수도 있다.
