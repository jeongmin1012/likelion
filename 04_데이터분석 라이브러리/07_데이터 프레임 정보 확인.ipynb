### 데이터의 종류
- 범위형 : 연속성을 가지고 있는 값을 의미한다. 최대나 최소가 있을 수 있다 (나이, 키, 몸무게 등등)
- 범주형 : 값의 종류가 정해져 있고 연속성을 가지고 있지 않는 데이터(성별, 혈액형, 지하철 노선 등등)

```python
# 요약 통계 정보
# describe : 숫자와 숫자가 아닌 컬럼들이 같이 있을 경우 숫자 컬럼들만 모아서 요약 통계 정보를 구해 반환한다.
# 만약 문자열만 있는 데이터 프레임 이라면 그에 맞는 요약 통계 정보를 제공한다.

# count : 결측치를 제외한 데이터의 개수
# mean : 결측치를 제외한 평균
# std : 결측치를 제외한 표준 편차
# min : 결측치를 제외한 최소 수
# 25% : 결측치를 제외한 25% 위치의 수
# 50% : 결측치를 제외한 50% 위치의 수
# 75% : 결측치를 제외한 75% 위치의 수
# max : 결측치를 제외한 최대 수
df1.describe()
```

```python
# 범위형 데이터만 추출하여 요약 통계정보를 확인한다.
df1[['SibSp', 'Parch', 'Fare']].describe()
```

```python
# 범주형 데이터를 통한 요약 통계 정보 확인
# 만약 숫자 데이터가 저장되어 있는 컬럼이 있다면 문자열 컬럼으로 변환해서 보세요
# count : 결측치를 제외한 데이터의 개수
# unique : 중복을 제외한 데이터의 종류 수
# top : 빈도수가 가장 높은 데이터
# freq : top에 해당하는 데이터가 몇 번이나 나왔는지..
df1[['Survived', 'Pclass', 'Sex']].astype('str').describe()
```

```python
# 범위형 데이터에 대한 요약 통계 정보에서 min과 max만 보고 데이터 명세서에 나와있는
# 정상 범위안에 최소와 최대값이 포함되는지 확인 한다.
a1 = df1[['SibSp', 'Parch', 'Fare']].describe()
a1.loc[['min', 'max']]
```

```python
# 범주형 데이터는 어떤한 값들이 저장되어 있는지 확인하는 것이 중요하다
# 하지만 요약 통계 정보를 통해서는 이를 확인하기가 어렵다.
# 각 컬럼별 중복을 제외한 데이터가 무엇이 있는지 직접 확인해야 한다
a1 = df1['Survived'].value_counts().index
display(a1)

a2 = df1['Survived'].unique()
display(a2)
```

```python
a1 = df1['Embarked'].value_counts().index
display(a1)

a2 = df1['Embarked'].unique()
display(a2)
```
