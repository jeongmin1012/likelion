### matplotlib
- 파이썬에서 데이터를 시각화 하기 위해 만든 라이브러리
- 다른 시각화 라이브러리들은 대부분 matplotlib를 통해 만들어진 라이브러리이다
- https://matplotlib.org/stable/gallery/index

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
# 그래프를 그리기 위한 기본 설정
#plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

### 그래프 그리기
- 라인 프롯을 통해 그래프를 그리는 기본 방법에 대해서 살펴본다.
- matplotlib는 그래프의 종류에 따라 함수가 제공되어 원하는 그래프를 그릴 수 있는 함수를 사용하면 된다.

```python
# x축에 적용할 데이터와 y축에 적용할 데이터를 준비하여 그래프를 그리는 함수를 호출하면
# 호출한 함수에 해당하는 그래프가 그려진다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
# 그래프를 그리는 함수들 일부는 축 하나의 값만 지정해도 되는 것들이 있다.
# 값 하나만 지정을 하면 y축에 적용하는 값이 되며 x 축을 알아서 설정된다.
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 일부 그래프를 그리는 함수들은 반드시 x와 y축의 값을 지정해야 하는 것도 있다.
plt.scatter([1, 4, 9, 16])
plt.show()
```

### 그래프에 적용할 수 있는 다양한 옵션들

```python
# 그래프의 제목
plt.title('그래프의 제목')

plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 제목의 위치를 설정한다.
# loc : left, right, center(기본)
plt.title('그래프의 제목', loc='left')

plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 제목의 세로방향 위치를 설정한다.
# y : 제목의 세로방향 위치를 설정한다. 0 이면 그래프의 아래쪽이된다.
# 1은 그래프의 y축 높이 만큼 된다. 기본은 None으로 설정되어 있으며 None은 1을 의미한다.
# loc는 가로방향 위치를 설정하는 옵션이다.
plt.title('그래프의 제목', y=1.2)

plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 각 축의 이름
plt.title('각 축의 이름')
plt.plot([1, 4, 9, 16])

# 각 축에 설정할 수 있는 옵션
# rotation : 회적 각도, 시계 반대 방향
# loc : 위치
# x축에 적용 시 : left, center, right
# y축에 적용 시 : top, center, bottom

# x축의 이름
plt.xlabel('X 축', loc='left')
# y축의 이름
plt.ylabel('Y 축', rotation=0, loc='bottom')

plt.show()
```

```python
# 전체적인 기본 설정
# 이 파일 상단에 미리 해두었습니다
# 윈도우 기본 폰트
# plt.rcParams['font.family'] = 'Malgun Gothic'
# 맥 기본 폰트
# plt.rcParams['font.family'] = 'AppleGothic'
# 글자 크기
# plt.rcParams['font.size'] = 14
# 한글 폰트를 적용하면 마이너스 기호가 깨진다.
# 마이너스 기호는 폰트를 적용하지 말라고 하는 설정
# plt.rcParams['axes.unicode_minus'] = False
```

```python
# 외부 폰트를 사용하는 방법은 다양한다.
# 그 중, 폰트 파일의 경로와 파일명을 지정하여 폰트를 사용하는 방법을 사용한다.
# 1. ipynb 파일이 있는 곳에 font 라는 폴더를 만들어준다.
# 2. 이 폴더에 사용하고자 하는 폰트 파일들을 모두 복사해준다.

# 사용할 폰트의 경로를 변수에 담아준다.
font_path1 = 'font/NanumPen.ttf'
# 폰트의 정보를 가지고 있는 객체를 생성한다.
font_prop1 = fm.FontProperties(fname=font_path1)
print(font_prop1.get_name())

font_path2 = 'font/NanumMyeongjo.ttf'
font_prop2 = fm.FontProperties(fname=font_path2)
print(font_prop2.get_name())

font_path3 = 'font/BagelFatOne-Regular.ttf'
font_prop3 = fm.FontProperties(fname=font_path3)
print(font_prop3.get_name())
```

```python
# 폰트 적용
# 문자열을 지정하는 함수들의 fontproperties라는 매개변수에 폰트를 관리하는 객체를 넣어준다.
plt.title('각 축의 이름', fontproperties=font_prop1)
plt.plot([1, 4, 9, 16])

plt.xlabel('X 축', fontproperties=font_prop2)
plt.ylabel('Y 축', fontproperties=font_prop3)

plt.show()
```

### 그래프의 스타일
- 그래프의 스타일은 문자열로 지정한다.
- 문자열 양식은 '색상 마커종류 선종류'형태로 작성한다(rs--)
---
- 색상
- https://matplotlib.org/stable/gallery/color/named_colors.html
- 위 주소에서 base color를 사용한다
---
- 마커
- https://matplotlib.org/stable/api/markers_api.html
---
- 선종류
- '-' : 실선
- '--' : dash
- '-.' : 점과 dash
- ':' : 점선
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
plt.title('그래프의 스타일')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.show()
```

```python
plt.title('그래프의 스타일')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'm*:')
plt.show()
```

```python
# 상세설정
# c : 선 색상
# lw : 선 굵기
# ls : 선 스타일
# marker : 마커 종류
# ms : 마커 크기
# mec : 마커 색성
# mew : 마커 선 굵기
# mfc : 마커 내부 색상

plt.title('그래프의 스타일')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='aqua', lw=5, ls=':', marker='o', ms=15, mec='g', mew=5, mfc='b')
plt.show()
```

### 기타 옵션들

```python
# 그리드 설정
plt.title('그리드')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
# 그리드를 표시한다.
plt.grid()
# 옛날 버전의 matplotlib는 grid가 표시되는 것이 기본이였다.
# 만약 이럴 경우 grid를 표시하고 싶지 않다면...
# plt.grid(False)
plt.show()
```

```python
plt.title('다수의 그래프')
# plt.show() 호출하기 전까지 그려진 그래프들은 모두 하나의
# 좌표 평면에 그려진다.
# 이 때, 각 그래프의 색상도 다르게 설정된다.
# 하지만 너무 많이 그리면 결국 같은 색상의 그래프들이 있을 수 있다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.plot([40, 30, 20, 10], [1, 4, 9, 16])
plt.show()
```

```python
# 서로 다른 좌표 평면에 그래프를 그린다.
plt.title('서로 다른 좌표 평면의 그래프')
# plt.show()를 호출하기 전까지만 하나의 좌표 평면에 그려진다.
# 중간 중간에 plt.show()를 호출하면 각각 서로 다른 좌표 평면에 그려진다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
plt.plot([40, 30, 20, 10], [1, 4, 9, 16])
plt.show()
```

```python
# subplot : 하나의 좌표평면 내부에 좌표 평면들을 배치해서 각각 그래프를
# 그릴 수 있도록 해준다.

# 방법1. 미리 나눠놓고 그래들을 그리는 방법
# 2행 2열
fig, ax = plt.subplots(2, 2)

ax[0][0].plot([10, 20, 30, 40], [1, 4, 9, 16])
ax[0][0].set_title('1행 1열')

ax[0][1].plot([10, 20, 30, 40], [1, 4, 9, 16])
ax[0][1].set_title('1행 2열')

ax[1][0].plot([10, 20, 30, 40], [1, 4, 9, 16])
ax[1][0].set_title('2행 1열')

ax[1][1].plot([10, 20, 30, 40], [1, 4, 9, 16])
ax[1][1].set_title('2행 2열')

# 그래프 간의 높이를 설정한다.
plt.subplots_adjust(hspace=0.4)

# 전체 타이틀
# y : 제목의 세로 높이(0은 제일 밑, 1은 제일 위)
plt.suptitle('Sub Plot 1', y=1.1)

plt.show()
```

```python
# 방법2
# 그래프를 그릴때 몇행 몇열 중 몇번째임을 지정하면서 그려준다.

# 2행 2열 중 첫 번째
plt.subplot(2, 2, 1)
plt.title('1행 1열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 2행 2열 중 두 번째
plt.subplot(2, 2, 2)
plt.title('1행 2열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 2행 2열 중 세 번째
plt.subplot(2, 2, 3)
plt.title('2행 1열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 2행 2열 중 네 번째
plt.subplot(2, 2, 4)
plt.title('2행 2열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 그래프 간의 높이를 설정한다.
plt.subplots_adjust(hspace=0.4)

# 전체 타이틀
# y : 제목의 세로 높이(0은 제일 밑, 1은 제일 위)
plt.suptitle('Sub Plot 1', y=1.1)

plt.show()
```

```python
# 범례

plt.title('범례')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], label='선1')
plt.plot([40, 30, 20, 10], [1, 4, 9, 16], label='선2')

# 범례를 표시한다.
# loc : 범례의 위치, 생략시 0
# 0 : best, 가장 최적의 위치를 자동으로 선정한다.
# 1 : 우상단, 2 : 좌상단, 3 : 좌하단, 4 : 우하단
# 5 : 중앙우측, 6 : 중앙좌측, 7 : 중앙우측, 8 : 중앙하단
# 9 : 중앙 상단, 10 : 정중앙
plt.legend(loc=10)

plt.show()
```

```python
# 범례

plt.title('범례')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], label='선1')
plt.plot([40, 30, 20, 10], [1, 4, 9, 16], label='선2')

# bbox_to_anchor : 좌표평면 밖에 범례를 표시하고자 할 때 사용한다.
# x 축 비율, y 축 비율
# 비율 : 각 축의 전체 길이가 1
plt.legend(bbox_to_anchor=(1.02, 1))

plt.show()
```
