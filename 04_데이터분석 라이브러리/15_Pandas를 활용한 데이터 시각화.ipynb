### 다른 그래프 그리기
- plot 함수에 kind라는 매개변수가 있다. 
- kind 매개변수에 그리고자 하는 그래프의 이름을 문자열로 넣어주면 그 그래프가 그려진다.
- 일부 그래프들은 그래프를 그리는 함수를 따로 제공하기도 한다.
- bar, pie, hist, kde, box, scatter, area

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
iris
```

```python
# Series 하나의 데이터를 막대 그래프로 그린다.
# Series의 값 하나가 막대 하나가 된다.
# 값의 index가 x축의 눈금으로 반영된다
iris2 = iris.copy()
a1 = iris2['꽃받침길이'][:10]
a1.index = ['꽃1', '꽃2', '꽃3', '꽃4', '꽃5', '꽃6', '꽃7', '꽃8', '꽃9', '꽃10']
display(a1)

a1.plot(kind='bar')
plt.title('막대그래프')
plt.show()
```

```python
# 일부 그래프는 함수를 따로 제공하기도 한다.
a1.plot.bar()
plt.title('막대그래프')
plt.show()
```

```python
# 데이터 프레임 (컬럼이 2개 이상)을 통해 그래프를 그린다.
# 하나의 로우를 추출하여 로우 내의 컬럼별 막대 그래프를 그리고
# 그것들을 하나의 막대 그룹으로 묶어준다.
# index가 x축의 눈금에 반영된다.
df2 = iris.drop('품종', axis=1)
a1 = df2.head(10)
a1.index = ['꽃1', '꽃2', '꽃3', '꽃4', '꽃5', '꽃6', '꽃7', '꽃8', '꽃9', '꽃10']
display(a1)

a1.plot(kind='bar')
plt.show()
```

### 파이 그래프
- Series를 통해 그린다.
- 각 Series의 값을 통해 비율을 계산하여 파이 그래프를 그린다.
- 매개변수의 구성은 matplotlib를 살펴봤을 때의 파이 그래프와 동일하다.

```python
a1 = iris['품종'].value_counts()
display(a1)

# 매개변수에 들어가는 것들은 matplotlib의 pie 와 동일하다
a1.plot(kind='pie', autopct='%.2f%%', startangle=90)
plt.title('파이그래프')
plt.show()
```

### 히스토그램
- 주어진 데이터 프레임에서 숫자로 구성되어 있는 컬럼들을 모아 히스토그램을 그려준다.
- 다수의 컬럼을 가지고 그리면 가려지는 부분들이 생기기 때문에 보기가 어려워진다.

```python
# bins : 구간의 개수
iris['꽃받침길이'].plot(kind='hist', bins=30)
plt.show()
```

### KDE
- 히스토그램은 다수의 컬럼을 지정하여 그릴 경우 중첩되는 부분이 생기게 된다.
- KDE는 히스토그램과 비슷하지만 막대가 아닌 선 그래프로 그린다.
- 히스토그램은 각 구간별로 양을 비교하지만 KDE는 밀도를 비교한다.

```python
a1 = iris.drop('품종', axis=1)
a1.plot(kind='hist', alpha=0.5)
plt.show()
```

```python
# scipy 라이브러리가 필요하다.
# pip install scipy
a1.plot(kind='kde')
plt.show()
```

### BoxPlot
- 값의 경계나 극단치 확인을 위해서 사용한다.
- IQR 값을 통해 극단치 최대 경계와 최소 경계를 계산하여 그 범위를 벗어나는 값은 원으로 표시해준다.

![](image/boxplot.jpg)

```python
# 범위형 데이터만 모아서 boxplot을 그린다.
a1 = iris.drop('품종', axis=1)

a1.plot(kind='box')
plt.show()
```

```python
# Q1(25%) 위치의 값과 Q3(75%위치의 값)
Q1 = iris['꽃받침너비'].quantile(q=0.25)
Q3 = iris['꽃받침너비'].quantile(q=0.75)
# IQR을 구한다.
IQR = Q3 - Q1
# 극단치 최소 경계
outliner_min = Q1 - (1.5 * IQR)
# 극단치 최대 경계
outliner_max = Q3 + (1.5 * IQR)

a1 = iris.query('꽃받침너비 < @outliner_min or 꽃받침너비 > @outliner_max')
a1
```

```python
# 범위형 데이터만 모아서 boxplot을 그린다.
a1 = iris.drop('품종', axis=1)
# boxplot 함수를 사용하면 똑같이 boxplot을 그릴 수 있는데 그리드가 기본적으로 표시된다.
a1.boxplot()
plt.show()
```
