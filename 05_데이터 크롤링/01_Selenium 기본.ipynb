```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

```python
# 웹드라이버 실행
driver = webdriver.Chrome()
```

```python
# 네이버 사이트를 요청한다.
driver.get('https://www.naver.com')
```

### 데이터 수집을 위해 반드시 알아야할 HTML 요소
- div : HTML 문서에서 영역을 구분하는 용도로 사용하는 태그
- nth-child : 동일한 경로의 태그들 중에 몇 번째 태그야~ 라는 의미
- h1 ~ h6 : 웹 문서에서 제목을 의미하는 태그이다.
- p : 웹 문서에서 단락을 구분하는 용도로 사용한다. div 태그와 비슷한 용도로 사용하기도 한다.
- span : 문자열을 묶어서 물자열에 대해 꾸며주는 용도로 사용한다. 문자열을 찾을 때 span 태그에 묶여 있는 경우가 많다.
- a : a태그로 묶여 있는 문자열이나 이미지 등을 클릭하면 a 태그가 가지고 있는 href 속성에 설정되어 있는 주소를 요청한다.
- ul, ol : html 문서 목록을 관리하는 용도로 사용하는 태그이다. 메뉴를 만들때도 사용한다.
- li : ul 또는 ol 안에서 항목 하나를 구성할 때 사용하는 태그이다. 메뉴라면 메뉴 항목 하나를 의미한다.

```python
# 회원 가입 태그 객체를 가져온다.
join_tag = driver.find_element(by=By.CSS_SELECTOR, value='#account > div > div > a:nth-child(3)')
# 태그를 클릭한다.
join_tag.click()
```

```python
# 약관 화면에서 "전체 동의하기"를 눌러준다.
check_tag = driver.find_element(by=By.CSS_SELECTOR, value='#join_form > div.form_content > div.form_section > div.check_terms > div > span')
check_tag.click()
```

```python
# 약관 화면에서 "전체 동의하기"를 눌러준다.
agree_btn = driver.find_element(by=By.CSS_SELECTOR, value='#btnAgree')
agree_btn.click()
```

```python
# 아이디 입력 칸에 아이디를 넣어준다
input_id = driver.find_element(by=By.CSS_SELECTOR, value='#id')
input_id.send_keys('woierhweorj')

# 비밀번호
input_pw = driver.find_element(by=By.CSS_SELECTOR, value='#pswd1')
input_pw.send_keys('weropwerpo')

# 이름
input_name = driver.find_element(by=By.CSS_SELECTOR, value='#name')
input_name.send_keys('박정민')

# 생년월일
input_birth = driver.find_element(by=By.CSS_SELECTOR, value='#birthdayInput')
input_birth.send_keys('20011012')

# 통신사
btn_telecom = driver.find_element(by=By.CSS_SELECTOR, value='#btn_telecom')
btn_telecom.click()

# 세부 통신사
btn_telecom_sub = driver.find_element(by=By.CSS_SELECTOR, value='#btnTelecom3')
btn_telecom_sub.click()
```
