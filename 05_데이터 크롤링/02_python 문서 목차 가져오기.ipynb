```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

```python
# 웹드라이버 실행
driver = webdriver.Chrome()
```

```python
# 사이트를 요청한다.
driver.get('https://docs.python.org/3/library/index.html')
```

```python
# ul 태그를 가져온다.
ul_tag = driver.find_element(by=By.CSS_SELECTOR, value='#the-python-standard-library > div > ul')
ul_tag
```

```python
# ul 태그 안에 있는 li 태그들을 가져온다.

# li 태그까지의 css selector를 가져오면 아래와 같다.
# li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='#the-python-standard-library > div > ul > li:nth-child(1)')

# 지금 ul_tag 안에서 태그를 찾는 것이기 때문에 ul 태그 까지의 selector인 "#the-python-standard-library > div > ul" 부분은
# 삭제한다
# li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='li:nth-child(1)')

# nth_child는 해당하는 태그 중 몇 번째 것이라는 의미이다. 지금은 다 가져올 것이 때문에 삭제한다.
li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='li')
li_list
```

```python
# ul 태그 안에 있는 li 태그들을 가져온다.

# li 태그까지의 css selector를 가져오면 아래와 같다.
# li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='#the-python-standard-library > div > ul > li:nth-child(1)')

# 지금 ul_tag 안에서 태그를 찾는 것이기 때문에 ul 태그 까지의 selector인 "#the-python-standard-library > div > ul" 부분은
# 삭제한다
# li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='li:nth-child(1)')

# nth_child는 해당하는 태그 중 몇 번째 것이라는 의미이다. 지금은 다 가져올 것이 때문에 삭제한다.
li_list = ul_tag.find_elements(by=By.CSS_SELECTOR, value='li > a')
# li_list
```

```python
# li_list의 수 만큼 만복한다.
for li_tag in li_list :
    print(li_tag.text)
```

```python
# a 태그의 문자열을 담을 리스트
text_list = []
# href 속성값을 담을 리스트
href_list = []

# li_list의 수 만큼 만복한다.
for li_tag in li_list :
    # 문자열 값을 담아준다
    text_list.append(li_tag.text)
    # href 속성 값을 담아준다.
    href_list.append(li_tag.get_attribute('href'))
```

```python
import numpy as np
import pandas as pd

a1 = {
    '학년' : [1, 2, 3],
    '나이' : [10, 20, 30],
    '이름' : ['홍길동', '김길동', '박길동']
}

a2 = {
    '학년' : [4, 5, 6],
    '나이' : [40, 50, 60],
    '이름' : ['최길동', '고길동', '황길동']
}

df1 = pd.DataFrame(a1)
df2 = pd.DataFrame(a2)

# 저장시 encoding을 utf-8-sig(웹 브라우저에서 사용하는 UTF-8 인코딩 셋)로 지정해주면
# 다음에 불러올때 encoding 설정을 안해도 되며 엑셀을 포함한 문서 도구들에서도 한글이 깨지지 않고 잘 나온다

# 최초로 저장하는 부분. index도 같이 저장되기 때문에 index=False를 넣어서 index는 저장되지 않게 한다.
df1.to_csv('test1.csv', index=False, encoding='utf-8-sig')
# 만약 파일에 추가로 저장하겠다면 mode='a' 를 넣어 추가될 수 있도록 해준다(생략하면 전의 내용을 모두 없애고 새롭게 저장한다)
# 컬럼 이름이 저장되는 것을 막기 위해 header=None이로 넣어준다.
df2.to_csv('test1.csv', index=False, encoding='utf-8-sig', mode='a', header=None)
```
