```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import numpy as np
import pandas as pd
```

```python
# 웹드라이버 실행
driver = webdriver.Chrome()
```

```python
# 사이트를 요청한다.
driver.get('https://namu.wiki/RecentChanges')
```

```python
div_tag = driver.find_element(by=By.CSS_SELECTOR, value='#app > div.H6qE-39I._4tQ0fN-2 > div._0Decl418._8wQSLjh0.yY1un6db > div > div.MAVxEdQh > article > div._85-dyOiX > div.rnAP1Jmw')

title_tag = div_tag.find_elements(by = By.CSS_SELECTOR, value = ':scope > div > div:nth-child(1)>a')
```

```python
for tag in title_tag :
    print(tag.text)
    print("------------------")
```

```python
title_list = []
history_list = []
compare_list = []
discuss_list = []

box_tag = div_tag.find_elements(by = By.CSS_SELECTOR, value = ':scope > div > div:nth-child(2)>div')

for tag,box in zip(title_tag,box_tag):
    title_list.append(tag.text)

    h = box.find_elements(by = By.CSS_SELECTOR, value = '[title="역사"]')
    c = box.find_elements(by = By.CSS_SELECTOR, value = '[title="비교"]')
    d = box.find_elements(by = By.CSS_SELECTOR, value ='[title="토론"]')

    history_list.append(h[0].tag_name == 'a')
    compare_list.append(c[0].tag_name == 'a')
    discuss_list.append(d[0].tag_name == 'a')


```

```python
df = pd.DataFrame({
    '제목': title_list,
    '역사': history_list,
    '비교': compare_list,
    '토론': discuss_list
})

df
```

```python
df.to_csv('namu_test.csv', index=False, encoding='utf-8-sig')
```
