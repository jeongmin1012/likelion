```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

import pandas as pd
```

```python
# 웹드라이버 실행
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
# 사이트를 요청한다.
driver.get('https://www.koreabaseball.com/Record/TeamRank/TeamRankDaily.aspx')
```

```python
# 팀 순위가 나와있는 table 태그내의 tr 태그들을 가져온다.
tr_list = driver.find_elements(by=By.CSS_SELECTOR, value='#cphContents_cphContents_cphContents_udpRecord > table > tbody > tr')
tr_list
```

```python
# 데이터를 담을 딕셔너리
d1 = {
    "순위" : [],
    "팀명" : [], 
    "경기" : [],
    "승" : [],
    "패" : [],
    "무" : [],
    "승률" : [],
    "게임차" : [],
    "최근10경기" : [],
    "연속" : [],
    "홈" : [],
    "방문" : []
}

# tr 태그의 수 만큼 반복한다.
for tr_tag in tr_list :
    # 현재 번째의 tr 태그 내의 tr 태그들을 가져온다.
    td_list = tr_tag.find_elements(by=By.CSS_SELECTOR, value='td')
    d1["순위"].append(td_list[0].text)
    d1["팀명"].append(td_list[1].text)
    d1["경기"].append(td_list[2].text)
    d1["승"].append(td_list[3].text)
    d1["패"].append(td_list[4].text)
    d1["무"].append(td_list[5].text)
    d1["승률"].append(td_list[6].text)
    d1["게임차"].append(td_list[7].text)
    d1["최근10경기"].append(td_list[8].text)
    d1["연속"].append(td_list[9].text)
    d1["홈"].append(td_list[10].text)
    d1["방문"].append(td_list[11].text)

df1 = pd.DataFrame(d1)
df1
```

```python
df1.to_csv('baseball.csv', encoding='utf-8-sig', index=False)
```
