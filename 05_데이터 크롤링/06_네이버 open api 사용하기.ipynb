### 문서를 보고 정리한 내용

1. 요청 주소
- https://openapi.naver.com/v1/search/news.json

2. 파라미터
- query : 저녁밥
- display : 100
- sort : date

3. 해더
- X-Naver-Client-Id : qlURw5tgE7CmccdW37fe
- X-Naver-Client-Secret : FOGoT6pEeT

4. 최종 요청 주소
- https://openapi.naver.com/v1/search/news.json?query=저녁밥&display=100&sort=date

```python
# pip install requests
import os
import requests
import json
```

```python
# 요청 주소
site = 'https://openapi.naver.com/v1/search/news.json?query=저녁밥&display=100&sort=date'
# 헤더로 보낼 데이터
headers = {
    'X-Naver-Client-Id' : 'qlURw5tgE7CmccdW37fe',
    'X-Naver-Client-Secret' : 'FOGoT6pEeT'
}
```

```python
# 서버로 부터 데이터를 받아온다.
response_data = requests.get(site, headers=headers)
print(response_data.text)
```

```python
# json 객체를 생성한다.
data_object = json.loads(response_data.text)
```

```python
for item in data_object['items'] :
    print(item['title'])
    print(item['originallink'])
    print(item['link'])
    print(item['description'])
    print(item['pubDate'])
    print()
```
