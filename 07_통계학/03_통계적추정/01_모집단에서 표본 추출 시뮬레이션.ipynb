```python
# 기본 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 그래프 기본설정\n",
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12,6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 다섯 마리의 물고기만 있는 호수에서 표본을 추출한다.
# 다음은 추출한 5마리 각각의 몸길이이다.
fish_5 = np.array([2, 3, 4, 5, 6])
fish_5
```

```python
# 무작위로 3마리를 추출한다.
# np.random.choice 함수 사용시 replace에 True를 넣어주면 추출한 값을 또다시 추출할 수 있다.
# False를 넣어주면 추출한 값은 다음 추출시 제외가 된다.
# 호수에서 물고기를 낚으면 그 물고기는 호수에서 없다는 가정을 한다.
np.random.seed(1)
sample_1 = np.random.choice(fish_5, size=3, replace=False)
sample_1
```

```python
# 골라낸 표본의 평균값을 계산한다
np.mean(sample_1)
```

```python
# 물고기 10만마리의 몸길이 데이터를 불러온다.
fish_100000 = pd.read_csv('data/5-2-1-fish_length_100000.csv')
fish_100000
```

```python
# 모집단에서 무작위로 500마리를 추출한다.
np.random.seed(1)

sample_2 = np.random.choice(fish_100000['length'], size=500, replace=False)
sample_2
```

```python
np.random.seed(1)

# 모집단에서 무작위로 10마리를 추출한다.
sample_10 = np.random.choice(fish_100000['length'], size=10, replace=False)
# 모집단에서 무작위로 100마리를 추출한다.
sample_100 = np.random.choice(fish_100000['length'], size=100, replace=False)
# 모집단에서 무작위로 1000마리를 추출한다.
sample_1000 = np.random.choice(fish_100000['length'], size=1000, replace=False)
# 모집단에서 무작위로 10000마리를 추출한다.
sample_10000 = np.random.choice(fish_100000['length'], size=10000, replace=False)
```

```python
# 평균을 구한다.
# 모집단 평균
a1 = np.mean(fish_100000['length'])
# 10마리 평균
a2 = np.mean(sample_10)
# 100마리 평균
a3 = np.mean(sample_100)
# 1000마리 평균
a4 = np.mean(sample_1000)
# 10000마리 평균
a5 = np.mean(sample_10000)

df1 = pd.DataFrame({
    '샘플수' : ['all', '10', '100', '1000', '10000'],
    '평균' : [a1, a2, a3, a4, a5]
})
df1
```

```python
# 분산을 구한다.
# 모집단 분산
a1 = np.var(fish_100000['length'])
# 10마리 분산
a2 = np.var(sample_10)
# 100마리 분산
a3 = np.var(sample_100)
# 1000마리 분산
a4 = np.var(sample_1000)
# 10000마리 분산
a5 = np.var(sample_10000)

df1 = pd.DataFrame({
    '샘플수' : ['all', '10', '100', '1000', '10000'],
    '분산' : [a1, a2, a3, a4, a5]
})
df1
```

```python
# 시각화한다.
sns.histplot(data=fish_100000, x='length')
plt.show()

sns.histplot(data=sample_10)
plt.show()

sns.histplot(data=sample_100)
plt.show()

sns.histplot(data=sample_1000)
plt.show()

sns.histplot(data=sample_10000)
plt.show()
```

```python
# 시각화한다.
plt.subplot(2, 3, 1)
sns.histplot(data=fish_100000, x='length')

plt.subplot(2, 3, 2)
sns.histplot(data=sample_10)

plt.subplot(2, 3, 3)
sns.histplot(data=sample_100)

plt.subplot(2, 3, 4)
sns.histplot(data=sample_1000)

plt.subplot(2, 3, 5)
sns.histplot(data=sample_10000)
plt.show()
```

```python
# 모집단의 확률 밀도를 구해본다.
a1 = np.mean(fish_100000['length'])
a2 = np.var(fish_100000['length'])
# 확률변수
x = np.arange(0, 8.1, 0.1)
# 확률밀도
density = stats.norm.pdf(x=x, loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
density_df
```

```python
# 모집단의 확률 밀도를 구해본다.
a1 = np.mean(fish_100000['length'])
a2 = np.var(fish_100000['length'])
# 확률변수
x = fish_100000['length']
# 확률밀도
density = stats.norm.pdf(x=fish_100000['length'], loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
density_df
```

```python
# 모집단 분포의 히스토그램
sns.histplot(fish_100000, stat='density')
# 확률 밀도
sns.lineplot(data=density_df, x='x', y='density')
plt.show()
```

```python
plt.subplot(2, 3, 1)
a1 = np.mean(fish_100000['length'])
a2 = np.var(fish_100000['length'])
x = fish_100000['length']
density = stats.norm.pdf(x=fish_100000['length'], loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
sns.histplot(fish_100000['length'], stat='density')
sns.lineplot(data=density_df, x='x', y='density')

plt.subplot(2, 3, 2)
a1 = np.mean(sample_10)
a2 = np.var(sample_10)
x = sample_10
density = stats.norm.pdf(x=sample_10, loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
sns.histplot(sample_10, stat='density')
sns.lineplot(data=density_df, x='x', y='density')

plt.subplot(2, 3, 3)
a1 = np.mean(sample_100)
a2 = np.var(sample_100)
x = sample_100
density = stats.norm.pdf(x=sample_100, loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
sns.histplot(sample_100, stat='density')
sns.lineplot(data=density_df, x='x', y='density')

plt.subplot(2, 3, 4)
a1 = np.mean(sample_1000)
a2 = np.var(sample_1000)
x = sample_1000
density = stats.norm.pdf(x=sample_1000, loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
sns.histplot(sample_1000, stat='density')
sns.lineplot(data=density_df, x='x', y='density')

plt.subplot(2, 3, 5)
a1 = np.mean(sample_10000)
a2 = np.var(sample_10000)
x = sample_10000
density = stats.norm.pdf(x=sample_10000, loc=a1, scale=a2)
density_df = pd.DataFrame({
    'x' : x,
    'density' : density
})
sns.histplot(sample_10000, stat='density')
sns.lineplot(data=density_df, x='x', y='density')

plt.show()
```

- 모집단이 정규분포를 이루고 있을 경우 무작위 샘플 추출을 많이 하면 모집단과 유사한 형태의 표본을 추출할 수 있다.
- 정균분포인 경우 대부분의 데이터가 평균에 가까이 있기 때문이다.
