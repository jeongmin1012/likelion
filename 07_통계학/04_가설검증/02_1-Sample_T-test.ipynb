### 1-sample T-test
- 단일 집단의 가설 검증시 사용한다.

```python
# 기본 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 그래프 기본설정\n",
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12,6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 가상의 데이터 생성
# 우리반 학생 30명의 평균 키가 173cm, 표준 편차가 5cm인 정규 분포를 따른다고 가정
np.random.seed(1)
my_class_heights = np.random.normal(173, 5, 30)
my_class_heights
```

```python
# 통계 검증 수행 (단일 표본 T-검정, 1-Sample T-test)
# stats.ttest_1samp(a=표본데이터, popmean=비교할 모집단 평균)
# 귀무가설(H0) : 우리 반 평균 키는 175cm와 같다 (차이가 없다)
# 대립가설(H1) : 우리 반 평균 키는 175cm와 다르다 (차이가 있다)
t_stat, p_val = stats.ttest_1samp(my_class_heights, 175)
print(f'p value : {p_val}')
```

```python
if p_val < 0.05 :
    print('귀무 가설을 기각한다')
else :
    print('귀무 가설을 채택한다')
```

```python
# 시각화
sns.histplot(my_class_heights, kde=True, color='skyblue', label='우리 반 데이터')
# 기준선 그리기 - 모집단 평균
plt.axvline(175, color='red', linestyle='--', linewidth=2, label='전국 평균 (175cm)')
# 실제값 그리기 - 표본 평균
a1 = np.mean(my_class_heights)
plt.axvline(x=a1, color='blue', linestyle='-', linewidth=2, label=f'우리반 평균 ({a1:.3f}cm)')

plt.title(f'단일 표본 T-test (p-vale : {p_val:.3f})')
plt.legend()
plt.show()
```

- 단일 표본 t-Test 검증 결과 p value가 0.05보다 작은 0.02 이므로 "우리반의 평균 키는 175cm와 같다"라는 귀무 가설을 기각한다.
- 따라서 "우리 반의 평균 키는 175cm와 다르다"
- 시각화를 통해 살펴본 결과 "우리반의 평균 키는 전국 평균인 175cm 보다 작다"
