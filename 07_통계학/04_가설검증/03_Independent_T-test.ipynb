### Independent T-test
- 독립된 두 집단을 비교하기 위해 사용한다.

```python
# 기본 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 그래프 기본설정\n",
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12,6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 생성
# np.random.normal(평균, 표준편차, 개수)
# A반 : 공부를 잘하는 편 (평균 80점, 표준편차 10점, 학생 30명)
# B반 : 성적이 조금 낮은 편(평균 70점, 표준편차 10점, 학생 30명)
class_a_score = np.random.normal(80, 10, 30)
class_b_score = np.random.normal(70, 10, 30)

print(f'A반 성적 평균 : {np.mean(class_a_score)}')
print(f'A반 표준 편차 : {np.std(class_a_score)}')
print(f'A반 학생 수 : {len(class_a_score)}')
print('-' * 30)
print(f'B반 성적 평균 : {np.mean(class_b_score)}')
print(f'B반 표준 편차 : {np.std(class_b_score)}')
print(f'B반 학생 수 : {len(class_b_score)}')
```

```python
# 독립 표본 T-검정
# stats.ttest_ind(집단1, 집단2)
# 귀무가설 : A반과 B반의 성적 평균은 같다.
# 대립가설 : A반과 B반의 성적 평균은 다르다.
t_stat, p_val = stats.ttest_ind(class_a_score, class_b_score)

print(f'p value : {p_val:.4f}')
```

```python
# 시각화를 위한 데이터 프레임 생성
df_ind = pd.DataFrame({
    # 1. 점수 데이터 [A반 점수들, B반 점수들]
    'Score' : np.concatenate([class_a_score, class_b_score]),
    # 2. 라벨 : 앞쪽 30개는 'A반', 뒤쪽 30개는 'B반'
    'Class' : ['A반'] * 30 + ['B반'] * 30
})
df_ind
```

```python
# 시각화를 위한 데이터 프레임 생성
df_ind = pd.DataFrame({
    # 1. 점수 데이터 [A반 점수들, B반 점수들]
    'Score' : np.concatenate([class_a_score, class_b_score]),
    # 2. 라벨 : 앞쪽 30개는 'A반', 뒤쪽 30개는 'B반'
    'Class' : ['A반'] * 30 + ['B반'] * 30
})
df_ind
```

```python
# 시각화
# 독립 표본 T-검정은 각 집단의 Boxplot으로 표현하여 비교를 한다.
# 박스 플롯을 통해 평균의 위치, 데이터의 분포 등을 비교한다.
sns.boxplot(data=df_ind, x='Class', y='Score', palette='Set2', hue='Class')
# 스트립 플롯
# 박스플롯만 보면 데이터 개수가 몇개인지 모르기 때문에 실제 점으로 표현해준다(그렇고 해서 데이터의 개수를 알 수 있는 것은 아니다.
# 실제 데이터 분포 정도만 확인 한다.
sns.stripplot(data=df_ind, x='Class', y='Score', color='black', alpha=0.3)

plt.title(f'독립 표본 T-test (p-vale={p_val:.4f})')
plt.show()
```

- 독립 표본 T-test 검증 결과 P value가 0.05보다 작은 0.001 이므로 "A반의 평균 성적과 B반의 평균 성적이 같다"라는 귀무 가설을 기각한다.
- 따라서 "A반의 평균 성적과 B반의 평균 성적은 다르다"
- 시각화를 통해 살펴본 결과 "A반의 평균 성적이 B반의 평균 성적보다 다 크다"
