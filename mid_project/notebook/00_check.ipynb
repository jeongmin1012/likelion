```python
team_war = bat.groupby(['Team','Year'])['WAR'].sum()
top3_war = (
    bat.sort_values('WAR', ascending=False)
       .groupby(['Team','Year']).head(3)
       .groupby(['Team','Year'])['WAR'].sum()
)
share = top3_war / team_war
```
