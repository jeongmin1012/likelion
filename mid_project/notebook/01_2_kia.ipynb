```python
import pandas as pd
```

```python
# 각각 season별 데이터 불러오기
bat = pd.read_csv('../data/kbo_batting_stats_by_season_1982-2025.csv')
pitch = pd.read_csv('../data/kbo_pitching_stats_by_season_1982-2025.csv')
```

```python
kia_bat = bat[
    (bat['Team'] == 'KIA') &
    (bat['Year'].between(2023, 2025))
]
```

```python
# 타석, 게임, 출루율, 장타율, 홈런, 볼넷, 유효타석당 득점, 대체선수 대비 승리기여도, 조정 득점 창출력, 타율, 병살타, 타수, 안타, 삼진
cols = ['PA','G','OBP','SLG','HR','BB','R/ePA','WAR','wRC+','AVG','GDP','AB','H','SO']
```

```python
kia_bat = kia_bat.groupby('Year')[cols].describe()
kia_bat.to_csv('kia_bat_yearly_describe.csv', encoding='utf-8-sig')
```

```python
kia_pitch = pitch[
    (pitch['Team'] == 'KIA') &
    (pitch['Year'].between(2023, 2025))
]
```

```python
cols = ['G', 'IP', 'TBF', 'GS', 'GR', 'ERA', 'WHIP', 'WAR','FIP', 'SO','BB','W','L','S','HD','H']
```

```python
kia_pitch = kia_pitch.groupby('Year')[cols].describe()
kia_pitch.to_csv('kia_pitch_yearly_describe.csv', encoding='utf-8-sig')
```
