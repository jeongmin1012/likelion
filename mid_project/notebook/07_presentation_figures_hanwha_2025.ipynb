```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 0) 설정
# -----------------------------
PATH = "../data/team_rank_2021_2025_10teams.csv"  # 네가 저장한 경로로 맞추기
YEARS = [2021,2022, 2023, 2024, 2025]
TARGET_TEAMS = ["한화", "LG", "KIA", "SSG"]  # 4팀만 보고 싶으면

TEAM_COLORS = {
    '한화': '#FC4E00',
    'LG': '#C30452',
    'KIA': '#EA0029',
    'SSG': '#CE0E2D',
    '삼성':'#074CA1',
    'NC':'#315288',
    'KT':'#000000',
    '롯데':'#041E42',
    '두산':'#1A1748',
    '키움':'#570514'
}

# -----------------------------
# 1) 데이터 로드 + 타입 정리
# -----------------------------
df = pd.read_csv(PATH, encoding="utf-8-sig")

# 혹시 문자열로 들어왔으면 숫자 변환(이미 변환해서 저장했어도 안전장치)
df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
df["승률"] = pd.to_numeric(df["승률"], errors="coerce")

# 분석/시각화에 필요한 행만
df = df[df["year"].isin(YEARS)].copy()

# -----------------------------
# 2) (옵션) 4팀만 필터
# -----------------------------
df_4 = df[df["팀명"].isin(TARGET_TEAMS)].copy()

# -----------------------------
# 3) 팀별 4년 승률 변화 라인차트
# -----------------------------
plt.figure(figsize=(10, 5))

for team, g in df_4.groupby("팀명"):
    g = g.sort_values("year")
    plt.plot(
        g["year"], g["승률"],
        marker="o",
        linewidth=2.5,
        label=team,
        color=TEAM_COLORS.get(team, None)
    )

plt.title("KBO 팀 승률 변화 (2022~2025) - LG,SSG,KIA,한화")
plt.xlabel("Year")
plt.ylabel("Win Rate (승률)")
plt.xticks(YEARS)
plt.grid(True, alpha=0.3)
plt.legend(title="Team")
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------
# 팀-연도 공격 지표 (PA >= 50 컷 적용)
# - 입력: kbo_batting_stats_by_season_1982-2025.csv
# - 출력: ../data/team_offense_2021_2025_pa50.csv
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import os

INPUT_PATH = "../data/raw/kbo_batting_stats_by_season_1982-2025.csv"
OUTPUT_PATH = "../data/team_offense_2021_2025_pa50.csv"

YEARS = [2021, 2022, 2023, 2024, 2025]
PA_CUT = 50

# 1) 로드
bat = pd.read_csv(INPUT_PATH)

# 2) 연도 필터
bat = bat[bat["Year"].isin(YEARS)].copy()

# 3) 숫자형 변환
num_cols = ["PA","AB","H","2B","3B","HR","TB","BB","HP","SO","SF","wRC+"]
for c in num_cols:
    if c in bat.columns:
        bat[c] = pd.to_numeric(bat[c], errors="coerce")

# 4) ✅ PA 컷 적용 (선수 단위)
bat = bat[bat["PA"] >= PA_CUT].copy()

# 5) 팀-연도 집계 함수
def team_agg(df: pd.DataFrame) -> pd.Series:
    PA = df["PA"].sum()
    AB = df["AB"].sum()
    H  = df["H"].sum()
    BB = df["BB"].sum()
    HBP = df["HP"].sum()
    SO = df["SO"].sum()
    TB = df["TB"].sum()
    SF = df["SF"].sum()

    # 팀 합계 기반 비율 지표
    AVG = H / AB if AB else np.nan
    OBP = (H + BB + HBP) / (AB + BB + HBP + SF) if (AB + BB + HBP + SF) else np.nan
    SLG = TB / AB if AB else np.nan
    ISO = SLG - AVG if (not np.isnan(SLG) and not np.isnan(AVG)) else np.nan

    SO_PA = SO / PA if PA else np.nan
    BBHBP_PA = (BB + HBP) / PA if PA else np.nan

    # 팀 wRC+ : PA 가중 평균
    mask = df["wRC+"].notna() & df["PA"].notna()
    wRC_plus = np.average(df.loc[mask, "wRC+"], weights=df.loc[mask, "PA"]) if mask.any() else np.nan

    return pd.Series({
        "PA": PA, "AB": AB, "H": H, "TB": TB,
        "BB": BB, "HBP": HBP, "SO": SO, "SF": SF,
        "AVG": AVG, "OBP": OBP, "SLG": SLG, "ISO": ISO,
        "SO_PA": SO_PA, "BBHBP_PA": BBHBP_PA,
        "wRC_plus": wRC_plus
    })

# 6) 팀-연도 집계
team_off = (
    bat.groupby(["Year","Team"], as_index=False)
       .apply(team_agg)
       .reset_index(drop=True)
       .rename(columns={"Year":"year", "Team":"팀명"})
       .sort_values(["year","팀명"])
       .reset_index(drop=True)
)

# 7) 저장
os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
team_off.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

print("✅ 저장 완료:", OUTPUT_PATH)
display(team_off.head())

# (확인용) 한화만
display(
    team_off[team_off["팀명"]=="한화"]
    [["year","wRC_plus","OBP","ISO","SO_PA","BBHBP_PA","PA"]]
)
```

```python
# ------------------------------------------------------------
# 팀-연도 공격 지표 (컷 없음)
# - 입력: kbo_batting_stats_by_season_1982-2025.csv
# - 출력: ../data/team_offense_2021_2025_nocut.csv
# - 포함: wRC_plus(PA가중평균), OBP, ISO, SO_PA, BBHBP_PA ...
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import os

INPUT_PATH = "../data/raw/kbo_batting_stats_by_season_1982-2025.csv"
OUTPUT_PATH = "../data/team_offense_2021_2025_nocut.csv"

YEARS = [2021, 2022, 2023, 2024, 2025]

# 1) 로드
bat = pd.read_csv(INPUT_PATH)

# 2) 연도 필터
bat = bat[bat["Year"].isin(YEARS)].copy()

# 3) 숫자형 변환
num_cols = ["PA","AB","H","2B","3B","HR","TB","BB","HP","SO","SF","wRC+"]
for c in num_cols:
    if c in bat.columns:
        bat[c] = pd.to_numeric(bat[c], errors="coerce")

# ✅ 4) 컷 없음: 여기서 필터링 안 함

def team_agg(df: pd.DataFrame) -> pd.Series:
    PA = df["PA"].sum()
    AB = df["AB"].sum()
    H  = df["H"].sum()
    BB = df["BB"].sum()
    HBP = df["HP"].sum()
    SO = df["SO"].sum()
    TB = df["TB"].sum()
    SF = df["SF"].sum()

    AVG = H / AB if AB else np.nan
    OBP = (H + BB + HBP) / (AB + BB + HBP + SF) if (AB + BB + HBP + SF) else np.nan
    SLG = TB / AB if AB else np.nan
    ISO = SLG - AVG if (not np.isnan(SLG) and not np.isnan(AVG)) else np.nan

    SO_PA = SO / PA if PA else np.nan
    BBHBP_PA = (BB + HBP) / PA if PA else np.nan

    # 팀 wRC+ : PA 가중 평균
    mask = df["wRC+"].notna() & df["PA"].notna() & (df["PA"] > 0)
    wRC_plus = np.average(df.loc[mask, "wRC+"], weights=df.loc[mask, "PA"]) if mask.any() else np.nan

    return pd.Series({
        "PA": PA, "AB": AB, "H": H, "TB": TB,
        "BB": BB, "HBP": HBP, "SO": SO, "SF": SF,
        "AVG": AVG, "OBP": OBP, "SLG": SLG, "ISO": ISO,
        "SO_PA": SO_PA, "BBHBP_PA": BBHBP_PA,
        "wRC_plus": wRC_plus
    })

# 5) 팀-연도 집계
team_off_nocut = (
    bat.groupby(["Year","Team"], as_index=False)
       .apply(team_agg)
       .reset_index(drop=True)
       .rename(columns={"Year":"year", "Team":"팀명"})
       .sort_values(["year","팀명"])
       .reset_index(drop=True)
)

# 6) 저장
os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
team_off_nocut.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

print("✅ 저장 완료:", OUTPUT_PATH)
display(team_off_nocut.head())

# (확인용) 한화만
display(
    team_off_nocut[team_off_nocut["팀명"]=="한화"]
    [["year","wRC_plus","OBP","ISO","SO_PA","BBHBP_PA","PA"]]
)
```

```python
# ------------------------------------------------------------
# 팀-연도 투수 고급 지표(FIP) 가공 (IP >= 30 컷)
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import os

INPUT_PATH = "../data/raw/kbo_pitching_stats_by_season_1982-2025.csv"
OUTPUT_PATH = "../data/team_pitcher_advanced_2021_2025_ip30.csv"

YEARS = [2021, 2022, 2023, 2024, 2025]
IP_CUT = 30

# 1) 로드
pit = pd.read_csv(INPUT_PATH)

# 2) 연도 필터
pit = pit[pit["Year"].isin(YEARS)].copy()

# 3) 숫자형 변환
num_cols = ["IP","HR","BB","HP","SO","ER"]
for c in num_cols:
    if c in pit.columns:
        pit[c] = pd.to_numeric(pit[c], errors="coerce")

# 4) IP 컷 (선수 단위)
pit = pit[pit["IP"] >= IP_CUT].copy()

# 5) 선수 단위 FIP (상수 제외)
pit["FIP_raw"] = (
    (13 * pit["HR"] + 3 * (pit["BB"] + pit["HP"]) - 2 * pit["SO"])
    / pit["IP"]
)

# 6) 팀-연도 집계
def team_pitcher_agg(df):
    IP = df["IP"].sum()
    HR = df["HR"].sum()
    BB = df["BB"].sum()
    HBP = df["HP"].sum()
    SO = df["SO"].sum()
    ER = df["ER"].sum()

    ERA = (ER * 9) / IP if IP else np.nan
    FIP = (13*HR + 3*(BB + HBP) - 2*SO) / IP if IP else np.nan

    return pd.Series({
        "IP": IP,
        "ERA": ERA,
        "FIP": FIP,
        "HR": HR,
        "BB": BB,
        "HBP": HBP,
        "SO": SO
    })

team_pit = (
    pit.groupby(["Year","Team"], as_index=False)
       .apply(team_pitcher_agg)
       .reset_index(drop=True)
       .rename(columns={"Year":"year","Team":"팀명"})
       .sort_values(["year","팀명"])
       .reset_index(drop=True)
)

# 7) 저장
os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
team_pit.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

print("✅ 저장 완료:", OUTPUT_PATH)
display(team_pit.head())

# (확인용) 한화만
display(
    team_pit[team_pit["팀명"]=="한화"]
    [["year","ERA","FIP","IP","SO","HR","BB","HBP"]]
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = "../data/processed/pitch_analysis_ip30_2021_2025.csv"
df = pd.read_csv(PATH)

TARGET_TEAM = "한화"
STRONG_TEAMS = ["LG", "KIA", "SSG"]
YEARS = [2021, 2022, 2023, 2024, 2025]
TEAM_GAMES = {2021:144, 2022:144, 2023:144, 2024:144, 2025:144}

TEAM_COLORS = { 
    '한화': '#FC4E00', 
    'LG': '#C30452', 
    'KIA': '#EA0029', 
    'SSG': '#CE0E2D',
    '삼성':'#074CA1', 
    'NC':'#315288', 
    'KT':'#000000', 
    '롯데':'#041E42', 
    '두산':'#1A1748',
    '키움':'#570514'
}
HANWHA_SUBCOLOR = "#2C3E50"  # 서브색(짙은 톤)

# -----------------------------
# 1) 기본 필터
# -----------------------------
df = df[df["Year"].isin(YEARS)].copy()
df = df.dropna(subset=["Year", "Team"])
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df[df["Year"].notna()]

# -----------------------------
# 2) 선발 평균 이닝 계산
#    - 우선순위: GS + IP가 있으면 선발 평균 이닝(=SP_IP/SP_GS)
#    - 없으면: 팀 IP / 경기수(144) 프록시
# -----------------------------
cols = set(df.columns)

has_gs = "GS" in cols
has_ip = "IP" in cols

if has_ip:
    df["IP"] = pd.to_numeric(df["IP"], errors="coerce")

if has_gs:
    df["GS"] = pd.to_numeric(df["GS"], errors="coerce")

if has_gs and has_ip:
    # ✅ 선발 등판이 있는 투수만 선발 풀로 간주 (GS>0)
    sp = df[(df["GS"].notna()) & (df["GS"] > 0) & (df["IP"].notna())].copy()

    team_year = (
        sp.groupby(["Year", "Team"], as_index=False)
          .agg(SP_IP=("IP", "sum"),
               SP_GS=("GS", "sum"))
    )
    team_year["sp_ip_per_start"] = team_year["SP_IP"] / team_year["SP_GS"]
    metric_col = "sp_ip_per_start"
    y_label = "선발 평균 이닝 (IP / GS)"
    title_suffix = " (IP≥30, 선발=GS>0 투수 IP/GS)"

else:
    # ✅ GS가 없으면 프록시: 팀 총 IP / 경기수
    if not has_ip:
        raise RuntimeError("이 파일에 IP 컬럼이 없어서 선발 평균 이닝(또는 프록시) 계산이 불가해요.")

    team_year = (
        df[df["IP"].notna()]
        .groupby(["Year", "Team"], as_index=False)
        .agg(TOTAL_IP=("IP", "sum"))
    )
    team_year["games"] = team_year["Year"].map(TEAM_GAMES)
    team_year["sp_ip_per_start"] = team_year["TOTAL_IP"] / team_year["games"]
    metric_col = "sp_ip_per_start"
    y_label = "경기당 평균 투구 이닝 (IP / 경기수)"
    title_suffix = " (IP≥30, 프록시=팀 IP/144)"

# -----------------------------
# 3) 비교선 만들기 (한화/리그평균/강팀평균)
# -----------------------------
hanwha = (
    team_year[team_year["Team"] == TARGET_TEAM]
    .set_index("Year")[[metric_col]]
    .rename(columns={metric_col:"Hanwha"})
)

league_avg = (
    team_year.groupby("Year")[metric_col].mean()
    .to_frame("League_Avg")
)

strong_avg = (
    team_year[team_year["Team"].isin(STRONG_TEAMS)]
    .groupby("Year")[metric_col].mean()
    .to_frame("Strong3_Avg")
)

compare_df = hanwha.join(league_avg).join(strong_avg)
print("\n[E-30: 연도별 선발 평균 이닝(또는 프록시)]")
print(compare_df.round(3))

# -----------------------------
# 4) 그래프
# -----------------------------
plt.figure(figsize=(9, 6))
plt.plot(compare_df.index, compare_df["Hanwha"],
         marker="o", linewidth=2.8, label="한화",
         color=TEAM_COLORS["한화"])
plt.plot(compare_df.index, compare_df["League_Avg"],
         marker="x", linestyle="--", linewidth=2.2, label="리그 평균",
         color=HANWHA_SUBCOLOR)
plt.plot(compare_df.index, compare_df["Strong3_Avg"],
         marker="s", linestyle=":", linewidth=2.2, label="강팀 3팀 평균",
         color="#666666")

for y in compare_df.index:
    plt.text(y, compare_df.loc[y, "Hanwha"], f"{compare_df.loc[y, 'Hanwha']:.2f}",
             ha="center", va="bottom", fontsize=9)

plt.title(f"한화 vs 리그 평균 vs 강팀 평균\n선발 평균 이닝{title_suffix}")
plt.xlabel("Year")
plt.ylabel(y_label)
plt.xticks(compare_df.index)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = "../data/processed/pitch_analysis_ip30_2021_2025.csv"
df = pd.read_csv(PATH)

TARGET_TEAM = "한화"
STRONG_TEAMS = ["LG", "KIA", "SSG"]
YEARS = [2021, 2022, 2023, 2024, 2025]

TEAM_COLORS = { 
    '한화': '#FC4E00', 
    'LG': '#C30452', 
    'KIA': '#EA0029', 
    'SSG': '#CE0E2D',
    '삼성':'#074CA1', 
    'NC':'#315288', 
    'KT':'#000000', 
    '롯데':'#041E42', 
    '두산':'#1A1748',
    '키움':'#570514'
}
HANWHA_SUBCOLOR = "#2C3E50"

# -----------------------------
# 1) 기본 필터
# -----------------------------
df = df[df["Year"].isin(YEARS)].copy()
df = df.dropna(subset=["Year", "Team"])
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df[df["Year"].notna()]

cols = set(df.columns)
has_gs = "GS" in cols
has_ip = "IP" in cols

if not (has_gs and has_ip):
    print("⚠️ 이 파일에는 GS(선발등판) 또는 IP 컬럼이 없어서 선발 이닝 비중(SP_IP / TOTAL_IP)을 계산할 수 없어요.")
    print("➡️ E-31은 생략하고, E-30(선발 평균 이닝) + E-32(GR 불펜 등판 수)로 운영 구조를 설명하는 걸 추천해요.")
else:
    df["GS"] = pd.to_numeric(df["GS"], errors="coerce")
    df["IP"] = pd.to_numeric(df["IP"], errors="coerce")

    # -----------------------------
    # 2) 팀-연도별 선발 이닝(SP_IP) / 전체 이닝(TOTAL_IP)
    # -----------------------------
    team_year_total = (
        df[df["IP"].notna()]
        .groupby(["Year","Team"], as_index=False)
        .agg(TOTAL_IP=("IP", "sum"))
    )

    sp = df[(df["GS"].notna()) & (df["GS"] > 0) & (df["IP"].notna())].copy()
    team_year_sp = (
        sp.groupby(["Year","Team"], as_index=False)
          .agg(SP_IP=("IP", "sum"))
    )

    team_year = team_year_total.merge(team_year_sp, on=["Year","Team"], how="left")
    team_year["SP_IP"] = team_year["SP_IP"].fillna(0)
    team_year["SP_share"] = team_year["SP_IP"] / team_year["TOTAL_IP"]

    # -----------------------------
    # 3) 비교선 만들기
    # -----------------------------
    hanwha = (
        team_year[team_year["Team"] == TARGET_TEAM]
        .set_index("Year")[["SP_share"]]
        .rename(columns={"SP_share":"Hanwha"})
    )

    league_avg = (
        team_year.groupby("Year")["SP_share"].mean()
        .to_frame("League_Avg")
    )

    strong_avg = (
        team_year[team_year["Team"].isin(STRONG_TEAMS)]
        .groupby("Year")["SP_share"].mean()
        .to_frame("Strong3_Avg")
    )

    compare_df = hanwha.join(league_avg).join(strong_avg)
    print("\n[E-31: 연도별 선발 이닝 비중(SP_IP / TOTAL_IP)]")
    print(compare_df.round(3))

    # -----------------------------
    # 4) 그래프
    # -----------------------------
    plt.figure(figsize=(9, 6))
    plt.plot(compare_df.index, compare_df["Hanwha"],
             marker="o", linewidth=2.8, label="한화",
             color=TEAM_COLORS["한화"])
    plt.plot(compare_df.index, compare_df["League_Avg"],
             marker="x", linestyle="--", linewidth=2.2, label="리그 평균",
             color=HANWHA_SUBCOLOR)
    plt.plot(compare_df.index, compare_df["Strong3_Avg"],
             marker="s", linestyle=":", linewidth=2.2, label="강팀 3팀 평균",
             color="#666666")

    for y in compare_df.index:
        plt.text(y, compare_df.loc[y, "Hanwha"], f"{compare_df.loc[y, 'Hanwha']:.2f}",
                 ha="center", va="bottom", fontsize=9)

    plt.title("한화 vs 리그 평균 vs 강팀 평균\n선발 이닝 비중 (SP_IP / TOTAL_IP, IP≥30)")
    plt.xlabel("Year")
    plt.ylabel("선발 이닝 비중")
    plt.xticks(compare_df.index)
    plt.ylim(0, 1)  # 비중이니까 0~1
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### E-32

```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = "../data/processed/pitch_analysis_ip30_2021_2025.csv"
df = pd.read_csv(PATH)

TARGET_TEAM = "한화"
STRONG_TEAMS = ["LG", "KIA", "SSG"]
YEARS = [2021, 2022, 2023, 2024, 2025]
TEAM_GAMES = {y:144 for y in YEARS}

TEAM_COLORS = { 
    '한화': '#FC4E00', 
    'LG': '#C30452', 
    'KIA': '#EA0029', 
    'SSG': '#CE0E2D',
    '삼성':'#074CA1', 
    'NC':'#315288', 
    'KT':'#000000', 
    '롯데':'#041E42', 
    '두산':'#1A1748',
    '키움':'#570514'
}
HANWHA_SUBCOLOR = "#2C3E50"

# -----------------------------
# 1) 기본 필터
# -----------------------------
df = df[df["Year"].isin(YEARS)].copy()
df = df.dropna(subset=["Year", "Team", "GR"])
df["GR"] = pd.to_numeric(df["GR"], errors="coerce")
df = df[df["GR"].notna() & (df["GR"] >= 0)]

# -----------------------------
# 2) 팀-연도별 불펜 등판 수
# -----------------------------
team_year = (
    df.groupby(["Year", "Team"], as_index=False)
      .agg(bullpen_appearances=("GR", "sum"))
)

team_year["games"] = team_year["Year"].map(TEAM_GAMES)
team_year["bp_per_game"] = team_year["bullpen_appearances"] / team_year["games"]

# -----------------------------
# 3) 비교선 구성
# -----------------------------
hanwha = (
    team_year[team_year["Team"] == TARGET_TEAM]
    .set_index("Year")[["bp_per_game"]]
    .rename(columns={"bp_per_game":"Hanwha"})
)

league_avg = (
    team_year.groupby("Year")["bp_per_game"].mean()
    .to_frame("League_Avg")
)

strong_avg = (
    team_year[team_year["Team"].isin(STRONG_TEAMS)]
    .groupby("Year")["bp_per_game"].mean()
    .to_frame("Strong3_Avg")
)

compare_df = hanwha.join(league_avg).join(strong_avg)
print("\n[E-32: 연도별 경기당 불펜 등판 수]")
print(compare_df.round(2))

# -----------------------------
# 4) 그래프
# -----------------------------
plt.figure(figsize=(9, 6))
plt.plot(compare_df.index, compare_df["Hanwha"],
         marker="o", linewidth=2.8,
         color=TEAM_COLORS["한화"], label="한화")
plt.plot(compare_df.index, compare_df["League_Avg"],
         marker="x", linestyle="--", linewidth=2.2,
         color=HANWHA_SUBCOLOR, label="리그 평균")
plt.plot(compare_df.index, compare_df["Strong3_Avg"],
         marker="s", linestyle=":", linewidth=2.2,
         color="#666666", label="강팀 3팀 평균")

for y in compare_df.index:
    plt.text(y, compare_df.loc[y, "Hanwha"],
             f"{compare_df.loc[y, 'Hanwha']:.2f}",
             ha="center", va="bottom", fontsize=9)

plt.title("한화 vs 리그 평균 vs 강팀 평균\n경기당 불펜 등판 수 (IP≥30, 불펜=GR)")
plt.xlabel("Year")
plt.ylabel("불펜 등판 수 / 경기")
plt.xticks(compare_df.index)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

### E-33

```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = "../data/processed/pitch_analysis_ip30_2021_2025.csv"
df = pd.read_csv(PATH)

TARGET_TEAM = "한화"
STRONG_TEAMS = ["LG", "KIA", "SSG"]
YEARS = [2021, 2022, 2023, 2024, 2025]

TEAM_COLORS = { 
    '한화': '#FC4E00', 
    'LG': '#C30452', 
    'KIA': '#EA0029', 
    'SSG': '#CE0E2D',
    '삼성':'#074CA1', 
    'NC':'#315288', 
    'KT':'#000000', 
    '롯데':'#041E42', 
    '두산':'#1A1748',
    '키움':'#570514'
}
HANWHA_SUBCOLOR = "#2C3E50"

# -----------------------------
# 1) 기본 필터
# -----------------------------
df = df[df["Year"].isin(YEARS)].copy()
df = df.dropna(subset=["Year", "Team", "GR"])
df["GR"] = pd.to_numeric(df["GR"], errors="coerce")
df = df[df["GR"].notna() & (df["GR"] >= 0)]

# 선발 제외 (불펜만)
if "GS" in df.columns:
    df = df[df["GS"] == 0]

# -----------------------------
# 2) 팀-연도별 핵심 불펜 집중도 계산
# -----------------------------
records = []

for (year, team), g in df.groupby(["Year", "Team"]):
    total_gr = g["GR"].sum()
    if total_gr == 0:
        continue

    top3_gr = (
        g.sort_values("GR", ascending=False)
         .head(3)["GR"]
         .sum()
    )

    records.append({
        "Year": year,
        "Team": team,
        "Top3_Share": top3_gr / total_gr
    })

focus_df = pd.DataFrame(records)

# -----------------------------
# 3) 비교선 만들기
# -----------------------------
hanwha = (
    focus_df[focus_df["Team"] == TARGET_TEAM]
    .set_index("Year")[["Top3_Share"]]
    .rename(columns={"Top3_Share":"Hanwha"})
)

league_avg = (
    focus_df.groupby("Year")["Top3_Share"].mean()
    .to_frame("League_Avg")
)

strong_avg = (
    focus_df[focus_df["Team"].isin(STRONG_TEAMS)]
    .groupby("Year")["Top3_Share"].mean()
    .to_frame("Strong3_Avg")
)

compare_df = hanwha.join(league_avg).join(strong_avg)
print("\n[E-33: 핵심 불펜 Top3 등판 집중도]")
print(compare_df.round(3))

# -----------------------------
# 4) 그래프
# -----------------------------
plt.figure(figsize=(9, 6))
plt.plot(compare_df.index, compare_df["Hanwha"],
         marker="o", linewidth=2.8,
         color=TEAM_COLORS["한화"], label="한화")
plt.plot(compare_df.index, compare_df["League_Avg"],
         marker="x", linestyle="--", linewidth=2.2,
         color=HANWHA_SUBCOLOR, label="리그 평균")
plt.plot(compare_df.index, compare_df["Strong3_Avg"],
         marker="s", linestyle=":", linewidth=2.2,
         color="#666666", label="강팀 3팀 평균")

for y in compare_df.index:
    plt.text(y, compare_df.loc[y, "Hanwha"],
             f"{compare_df.loc[y, 'Hanwha']:.2f}",
             ha="center", va="bottom", fontsize=9)

plt.title("한화 vs 리그 평균 vs 강팀 평균\n핵심 불펜 Top3 등판 집중도")
plt.xlabel("Year")
plt.ylabel("Top3 GR 비중")
plt.xticks(compare_df.index)
plt.ylim(0, 1)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
